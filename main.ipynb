```python
import os, re, json, gzip, random, time, textwrap, itertools, datetime as dt
from collections import defaultdict, Counter
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patheffects as pe
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.patches as mpatches
import matplotlib.dates as mdates
from matplotlib import animation
import warnings
warnings.filterwarnings("ignore")

# ───────────────────── 1.  Imports (heavy) ──────────────────────────
import pandas as pd
import numpy as np
import networkx as nx
import igraph as ig
from github import Github
from tqdm import tqdm
from community import community_louvain          # Louvain impl
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.manifold import TSNE, MDS
from sklearn.cluster import SpectralClustering
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import holoviews as hv
from holoviews import opts, dim
hv.extension('bokeh')

# Add new imports
from fa2_modified import ForceAtlas2
from textwrap import fill
from scipy.stats import powerlaw
from scipy.optimize import curve_fit
import nxviz as nv
import cdlib.algorithms as cdlib_algs
from cdlib import evaluation, NodeClustering
import powerlaw as pl
import matplotlib

# Optional imports with fallback
try:
    pass  # Placeholder to avoid syntax error
except Exception as e:
    print(f"An error occurred: {e}")
    import cairocffi as cairo
except ImportError:
    pass
    
try:
    from karateclub import Role2Vec, GraphWave
except ImportError:
    pass

try: 
    import statsmodels.api as sm
except ImportError:
    pass

# ───────────────────── 2.  Config ───────────────────────────────────
TOKEN = "YOUR_GITHUB_TOKEN"  # Replace with your GitHub token
HEADERS = {"Accept": "application/vnd.github+json"}
PALETTE = sns.color_palette("viridis", 10).as_hex()
TIME_WINDOWS = 6  # Number of time windows for temporal analysis

# ───────────────────── 3.  Helper: alias resolution ─────────────────
def canonical_login(user):
    """Collapse GitHub aliases by lower‑casing & stripping homographs."""
    if not user: return None
    login = user.login.lower() if hasattr(user, 'login') else str(user).lower()
    # Heuristically drop numerals if same core handle exists (e.g., 'alice' vs 'alice123')
    base = re.sub(r"\d+$", "", login)
    return base or login

# ───────────────────── 4.  Data Harvesters ──────────────────────────
from github.GithubException import UnknownObjectException

def fetch_user_edges_api(repo_fullname, max_pages=1, with_timestamps=True):
    """
    Enhanced GitHub REST harvester:
      • Commit author + committer
      • 'Co‑authored‑by:' footers (robust to names vs @handles)
      • PR‑review bridges
      • Returns timestamps for temporal analysis
    """
    gh = Github(TOKEN, per_page=100)
    repo = gh.get_repo(repo_fullname)
    edges = []

    co_footer = re.compile(
        r"Co-authored-by:\s+(.+?)\s*<(.+?)>", flags=re.I
    )  # captures "Name <email>"

    # Also gather user metadata for enrichment
    user_metadata = {}

    # —— (a) COMMITS
    for commit in tqdm(repo.get_commits()[: max_pages * 10],
                       desc=f"{repo_fullname} commits"):
        authors = set()
        commit_time = commit.last_modified

        # Convert string date to datetime
        if isinstance(commit_time, str):
            try:
                commit_time = dt.datetime.strptime(commit_time, "%a, %d %b %Y %H:%M:%S %Z")
            except:
                commit_time = dt.datetime.now()  # Fallback

        # primary author / committer
        for usr in (commit.author, commit.committer):
            if usr:
                login = canonical_login(usr)
                authors.add(login)
                
                # Collect metadata if not already present
                if login not in user_metadata and usr:
                    try:
                        gh_user = gh.get_user(login)
                        user_metadata[login] = {
                            'name': gh_user.name,
                            'company': gh_user.company,
                            'location': gh_user.location,
                            'email': gh_user.email,
                            'bio': gh_user.bio,
                            'created_at': gh_user.created_at,
                            'followers': gh_user.followers,
                            'public_repos': gh_user.public_repos
                        }
                    except:
                        # Skip if can't fetch user details
                        user_metadata[login] = {'name': login}

        # co‑author footers
        for name, email in co_footer.findall(commit.commit.message):
            # try to pull a GitHub handle out of the email (e.g. "login@users.noreply.github.com")
            m = re.match(r"^([a-zA-Z0-9-]+)@users\.noreply\.github\.com$", email)
            login_guess = m.group(1) if m else None
            if login_guess:
                authors.add(login_guess.lower())
            else:
                # final fallback: keep the name as a pseudo‑alias
                authors.add(name.lower().replace(" ", ""))

        # record pairwise edges
        for a, b in itertools.combinations(sorted(authors), 2):
            if with_timestamps:
                edges.append((a, b, commit_time))
            else:
                edges.append((a, b))

    # —— (b) PULL‑REQUEST REVIEWS
    for pr in tqdm(repo.get_pulls(state="closed")[:10], desc=f"{repo_fullname} PRs"):
        opener = canonical_login(pr.user)
        
        # Collect PR opener metadata
        if opener not in user_metadata and pr.user:
            try:
                gh_user = gh.get_user(opener)
                user_metadata[opener] = {
                    'name': gh_user.name,
                    'company': gh_user.company,
                    'location': gh_user.location,
                    'email': gh_user.email,
                    'bio': gh_user.bio,
                    'created_at': gh_user.created_at,
                    'followers': gh_user.followers,
                    'public_repos': gh_user.public_repos
                }
            except:
                user_metadata[opener] = {'name': opener}
        
        try:
            for rev in pr.get_reviews():
                reviewer = canonical_login(rev.user)
                if opener and reviewer and opener != reviewer:
                    if with_timestamps:
                        edges.append((opener, reviewer, rev.submitted_at))
                    else:
                        edges.append((opener, reviewer))
                        
                    # Collect reviewer metadata
                    if reviewer not in user_metadata and rev.user:
                        try:
                            gh_user = gh.get_user(reviewer)
                            user_metadata[reviewer] = {
                                'name': gh_user.name,
                                'company': gh_user.company,
                                'location': gh_user.location,
                                'email': gh_user.email,
                                'bio': gh_user.bio,
                                'created_at': gh_user.created_at,
                                'followers': gh_user.followers,
                                'public_repos': gh_user.public_repos
                            }
                        except:
                            user_metadata[reviewer] = {'name': reviewer}
        except UnknownObjectException:
            # a deleted account or inaccessible review – skip it
            continue

    return edges, user_metadata

# ───────────────────── 5.  Graph Builder ────────────────────────────
def build_graph(edge_list, min_interactions=2, with_timestamps=True):
    """
    Enhanced undirected, weighted contributor graph with edge weight = #co‑events.
    Returns both full temporal graph and filtered static graph.
    """
    G = nx.Graph()
    G_temporal = nx.Graph()
    
    if with_timestamps:
        # Create dataframe for temporal analysis
        edges_df = pd.DataFrame(edge_list, columns=['source', 'target', 'timestamp'])
        # Convert string dates to datetime if needed
        
        # if isinstance(edges_df['timestamp'].iloc[0], str):
        #     edges_df['timestamp'] = pd.to_datetime(edges_df['timestamp'])
        #     # Parse all timestamps as UTC (offset-aware), then sort
        
        edges_df['timestamp'] = pd.to_datetime(edges_df['timestamp'], utc=True)

        # Sort by timestamp
        edges_df = edges_df.sort_values('timestamp')
        
        # Add edges to both graphs
        for _, row in edges_df.iterrows():
            u, v, ts = row['source'], row['target'], row['timestamp']
            if u and v:
                # Add to static graph
                G.add_edge(u, v, weight=G[u][v]['weight']+1 if G.has_edge(u,v) else 1)
                
                # Add to temporal graph with timestamp
                if G_temporal.has_edge(u, v):
                    G_temporal[u][v]['timestamps'].append(ts)
                    G_temporal[u][v]['weight'] += 1
                else:
                    G_temporal.add_edge(u, v, timestamps=[ts], weight=1)
    else:
        # Simple edge list without timestamps
        for u, v in edge_list:
            if u and v:
                G.add_edge(u, v, weight=G[u][v]['weight']+1 if G.has_edge(u,v) else 1)
                G_temporal = G.copy()  # Just duplicate for consistency
    
    # Filter the static graph
    weak = [(u,v) for u,v,d in G.edges(data=True) if d['weight'] < min_interactions]
    G.remove_edges_from(weak)
    G.remove_nodes_from(list(nx.isolates(G)))
    
    return G, G_temporal, edges_df if with_timestamps else None

# ───────────────────── 6. Novel Network Metrics ────────────────────────────
def fit_power_law(degrees):
    """Fit a power-law distribution to degree data"""
    # Remove zeros which can't be fitted to power law
    degrees = [d for d in degrees if d > 0]
    if not degrees:
        return None, None
    
    # Fit using powerlaw package for more robust fitting
    results = pl.Fit(degrees)
    alpha = results.alpha
    xmin = results.xmin
    
    return alpha, xmin

def compute_rich_club_profile(G, normalized=True):
    """Get a more detailed rich club profile"""
    degrees = sorted(dict(G.degree()).values())
    if not degrees:
        return {}
        
    max_degree = max(degrees)
    k_values = range(1, min(max_degree + 1, 51))  # Up to max degree or 50
    
    try:
        rc = nx.rich_club_coefficient(G, normalized=normalized)
        # Fill in missing values with zeros
        return {k: rc.get(k, 0) for k in k_values}
    except:
        # Fallback for disconnected graphs
        return {k: 0 for k in k_values}

def temporal_network_evolution(edges_df, window_count=TIME_WINDOWS):
    """Analyze network evolution across time windows"""
    # Sort by timestamp
    edges_df = edges_df.sort_values('timestamp')
    
    # Define time windows
    min_time = edges_df['timestamp'].min()
    max_time = edges_df['timestamp'].max()
    window_size = (max_time - min_time) / window_count
    
    windows = []
    metrics_over_time = []
    
    for i in range(window_count):
        start_time = min_time + i * window_size
        end_time = start_time + window_size
        window_name = f"Window {i+1}"
        
        # Get edges in this window
        window_edges = edges_df[(edges_df['timestamp'] >= start_time) & 
                               (edges_df['timestamp'] < end_time)]
        
        if len(window_edges) > 0:
            # Build graph for this window
            edge_list = [(row['source'], row['target']) for _, row in window_edges.iterrows()]
            G_window = nx.Graph()
            for u, v in edge_list:
                if u and v:
                    G_window.add_edge(u, v, weight=G_window[u][v]['weight']+1 if G_window.has_edge(u,v) else 1)
            
            # Remove isolates
            G_window.remove_nodes_from(list(nx.isolates(G_window)))
            
            # Compute metrics
            if G_window.number_of_nodes() > 0:
                # Basic metrics
                n_nodes = G_window.number_of_nodes()
                n_edges = G_window.number_of_edges()
                density = nx.density(G_window)
                
                # Advanced metrics (safely)
                components = list(nx.connected_components(G_window))
                largest_cc = max(components, key=len) if components else []
                
                try:
                    avg_clustering = nx.average_clustering(G_window)
                except:
                    avg_clustering = 0
                
                try:
                    # Try to compute centrality measures on largest component
                    G_lcc = G_window.subgraph(largest_cc)
                    if len(G_lcc) > 0:
                        centrality = nx.eigenvector_centrality(G_lcc, max_iter=50)
                        top_node = max(centrality.items(), key=lambda x: x[1])[0] if centrality else None
                        top_centrality = centrality[top_node] if top_node else 0
                    else:
                        top_node = None
                        top_centrality = 0
                except:
                    top_node = None
                    top_centrality = 0
                
                # Community detection
                try:
                    partition = community_louvain.best_partition(G_window)
                    communities = len(set(partition.values()))
                    modularity = community_louvain.modularity(partition, G_window)
                except:
                    communities = 1
                    modularity = 0
                
                windows.append({
                    'window': window_name,
                    'start_time': start_time,
                    'end_time': end_time,
                    'graph': G_window,
                    'n_nodes': n_nodes,
                    'n_edges': n_edges,
                    'density': density,
                    'avg_clustering': avg_clustering,
                    'components': len(components),
                    'largest_cc_size': len(largest_cc),
                    'top_node': top_node,
                    'top_centrality': top_centrality,
                    'communities': communities,
                    'modularity': modularity
                })
                
                # Track specific metrics for plotting
                metrics_over_time.append({
                    'window': i+1,
                    'time': start_time + window_size/2,  # Middle of window
                    'nodes': n_nodes,
                    'edges': n_edges,
                    'density': density,
                    'avg_clustering': avg_clustering,
                    'communities': communities,
                    'modularity': modularity,
                })
    
    return windows, pd.DataFrame(metrics_over_time)

def novel_metrics(G):
    """
    Compute advanced metrics with focus on structural roles and community analyses
    """
    metrics = {}
    
    # Safe component handling
    components = list(nx.connected_components(G))
    largest_cc = max(components, key=len) if components else []
    G_lcc = G.subgraph(largest_cc)
    
    # —— Basic centrality metrics (safely executed) ——
    try:
        degree = dict(G.degree(weight='weight'))
    except:
        degree = {n: 0 for n in G.nodes()}
    
    try:
        betweenness = nx.betweenness_centrality(G, weight='weight')
    except:
        betweenness = {n: 0 for n in G.nodes()}
    
    try:
        closeness = nx.closeness_centrality(G)
    except:
        closeness = {n: 0 for n in G.nodes()}
    
    # —— Component-wise eigenvector centrality ——
    eig = {}
    for comp_nodes in nx.connected_components(G):
        sub = G.subgraph(comp_nodes)
        if len(sub) > 1:
            try:
                sub_eig = nx.eigenvector_centrality(sub, weight="weight", max_iter=1000)
                eig.update(sub_eig)
            except:
                # Fallback to degree centrality
                sub_eig = nx.degree_centrality(sub)
                eig.update(sub_eig)
        else:
            # Single node
            node = list(sub.nodes())[0]
            eig[node] = 0.0
    
    # —— Advanced structural metrics ——
    try:
        pagerank = nx.pagerank(G, weight="weight")
    except:
        pagerank = {n: 0 for n in G.nodes()}
    
    try:
        core = nx.core_number(G)
    except:
        core = {n: 0 for n in G.nodes()}
    
    try:
        clustering = nx.clustering(G, weight="weight")
    except:
        clustering = {n: 0 for n in G.nodes()}
    
    try:
        eff_size = nx.effective_size(G)
    except:
        eff_size = {n: 0 for n in G.nodes()}
    
    try:
        constraint = nx.constraint(G)
    except:
        constraint = {n: 0 for n in G.nodes()}
    
    # —— Novel: Structural Role Detection ——
    # Simplified role detection based on centrality patterns
    structural_roles = {}
    roles = {}
    
    for node in G.nodes():
        # Default role assignments
        deg = degree.get(node, 0)
        bet = betweenness.get(node, 0)
        eig_val = eig.get(node, 0)
        clust = clustering.get(node, 0)
        core_val = core.get(node, 0)
        
        # Normalize metrics for comparison
        max_bet = max(betweenness.values()) if betweenness.values() else 1
        max_eig = max(eig.values()) if eig.values() else 1
        max_core = max(core.values()) if core.values() else 1
        
        norm_bet = bet / max_bet if max_bet > 0 else 0
        norm_eig = eig_val / max_eig if max_eig > 0 else 0
        norm_core = core_val / max_core if max_core > 0 else 0
        
        # Classify nodes into structural roles
        if norm_eig > 0.7 and norm_core > 0.7:
            role = "Hub"  # High influence, central to network
        elif norm_bet > 0.5 and clust < 0.3:
            role = "Bridge"  # Connects different parts of network
        elif norm_core > 0.5 and clust > 0.5:
            role = "Core"  # Part of dense core community
        elif deg < 2:
            role = "Peripheral"  # On the edges of the network
        else:
            role = "Regular"  # Standard contributor
        
        # Store the role
        structural_roles[node] = role
        # Also store as a simple numeric value for plotting
        role_map = {"Hub": 0, "Bridge": 1, "Core": 2, "Regular": 3, "Peripheral": 4}
        roles[node] = role_map[role]
    
    # —— Assemble metrics dictionary ——
    for n in G.nodes():
        metrics[n] = {
            'degree': degree.get(n, 0),
            'weighted_degree': G.degree(n, weight='weight') if 'weight' in G[list(G.neighbors(n))[0]][n] else G.degree(n),
            'betweenness': betweenness.get(n, 0),
            'closeness': closeness.get(n, 0),
            'eigenvector': eig.get(n, 0),
            'pagerank': pagerank.get(n, 0),
            'coreness': core.get(n, 0),
            'clustering': clustering.get(n, 0),
            'effective_size': eff_size.get(n, 0),
            'constraint': constraint.get(n, 0),
            'role': roles.get(n, 0),
            'structural_role': structural_roles.get(n, "Regular"),
            'constraint_inv': 1 - constraint.get(n, 0),  # Structural holes metric
        }
        
        # Add "influencer score" - a weighted combination of normalized metrics
        # Scale each metric to [0,1] across network
        if G.nodes():
            # max_deg = max([metrics[node].get('weighted_degree', 0) for node in G.nodes()]) or 1
            # max_btw = max([metrics[node].get('betweenness',     0) for node in G.nodes()]) or 1
            # max_eig = max([metrics[node].get('eigenvector',    0) for node in G.nodes()]) or 1
            # max_pag = max([metrics[node].get('pagerank',       0) for node in G.nodes()]) or 1
            # max_cor = max([metrics[node].get('coreness',       0) for node in G.nodes()]) or 1
            # max_deg = max([metrics[node]['weighted_degree'] for node in G.nodes()]) or 1
            # max_btw = max([metrics[node]['betweenness'] for node in G.nodes()]) or 1
            # max_eig = max([metrics[node]['eigenvector'] for node in G.nodes()]) or 1
            # max_pag = max([metrics[node]['pagerank'] for node in G.nodes()]) or 1
            # max_cor = max([metrics[node]['coreness'] for node in G.nodes()]) or 1
            max_deg = max([metrics.get(node, {}).get('weighted_degree', 0) for node in G.nodes()]) or 1
            max_btw = max([metrics.get(node, {}).get('betweenness',      0) for node in G.nodes()]) or 1
            max_eig = max([metrics.get(node, {}).get('eigenvector',      0) for node in G.nodes()]) or 1
            max_pag = max([metrics.get(node, {}).get('pagerank',         0) for node in G.nodes()]) or 1
            max_cor = max([metrics.get(node, {}).get('coreness',         0) for node in G.nodes()]) or 1
            
            # Calculate influencer score (weighted combination)
            metrics[n]['influencer_score'] = (
                0.25 * (metrics[n]['weighted_degree'] / max_deg) +
                0.20 * (metrics[n]['betweenness'] / max_btw) +
                0.25 * (metrics[n]['eigenvector'] / max_eig) +
                0.20 * (metrics[n]['pagerank'] / max_pag) +
                0.10 * (metrics[n]['coreness'] / max_cor)
            )
            
            # Calculate structural bridge score (identifies important connectors)
            metrics[n]['bridge_score'] = (
                0.40 * (metrics[n]['betweenness'] / max_btw) +
                0.30 * (metrics[n]['constraint_inv']) +  # Already normalized
                0.30 * (1 - metrics[n]['clustering'])    # Invert clustering
            )
    

    # (optional) initialize ForceAtlas2 for later layout use
    try:
        forceatlas2 = ForceAtlas2(
            outboundAttractionDistribution=True,  # Dissuade hubs
            linLogMode=False,  # NOT LinLog mode
            adjustSizes=False,  # Prevent overlap
            edgeWeightInfluence=1.0,
            jitterTolerance=1.0,  # Tolerance
            barnesHutOptimize=True,
            barnesHutTheta=1.2,
            multiThreaded=False,  # NOT for jupyter
            scalingRatio=2.0,
            strongGravityMode=False,
            gravity=1.0,
            verbose=False
        )
    except Exception as e:
        print(f"Warning: Could not initialize ForceAtlas2: {e}")
        forceatlas2 = None

    # *** ADD THESE LINES TO RETURN THE THREE EXPECTED OBJECTS ***
    # 1) per-node metric dict
    node_metrics = metrics
    # 2) rich‐club profile
    richclub_data = compute_rich_club_profile(G, normalized=True)
    # 3) global network summary
    network_metrics = {
        'n_nodes': G.number_of_nodes(),
        'n_edges': G.number_of_edges(),
        'density': nx.density(G),
        'components': nx.number_connected_components(G),
        'largest_cc_size': max(len(c) for c in nx.connected_components(G)) if G.number_of_nodes()>0 else 0,
        'avg_clustering': nx.average_clustering(G) if G.number_of_nodes()>0 else 0,
        # add anything else you need, e.g. path length, diameter, power-law fit, etc.
    }



    # ─── Compute average shortest path length ─────────────────────────
    # Average shortest path length = sum of d(u,v) over all pairs, normalized by n*(n-1) :contentReference[oaicite:3]{index=3}
    try:
        network_metrics['avg_path_length'] = nx.average_shortest_path_length(G)
    except Exception:
        # Graph disconnected or too large; fallback to None
        network_metrics['avg_path_length'] = None

    # ─── Compute diameter ─────────────────────────────────────────────
    # Diameter = maximum eccentricity (longest shortest-path) :contentReference[oaicite:4]{index=4}
    try:
        network_metrics['diameter'] = nx.diameter(G)
    except Exception:
        # Graph disconnected or directed without strong connectivity
        network_metrics['diameter'] = None

    # ─── Compute degree assortativity ────────────────────────────────
    # Degree assortativity measures similarity in node-degree connections :contentReference[oaicite:5]{index=5}
    try:
        network_metrics['degree_assortativity'] = nx.degree_assortativity_coefficient(G)
    except Exception:
        network_metrics['degree_assortativity'] = None

    # ─── (Optional) add power-law fit if you haven’t already ─────────
    degrees = [d for _, d in G.degree()]
    alpha, xmin = fit_power_law(degrees)
    network_metrics['power_law_alpha'] = alpha
    network_metrics['power_law_xmin']  = xmin






    # ─── add power-law parameters ───────────────────────────
    degrees = [d for _, d in G.degree()]
    alpha, xmin = fit_power_law(degrees)
    network_metrics['power_law_alpha'] = alpha
    network_metrics['power_law_xmin']  = xmin
    # ─────────────────────────────────────────────────────────


    
    return node_metrics, richclub_data, network_metrics

def community_scan(G):
    """
    Run multiple community detection algorithms and evaluate their consistency
    """
    # Apply multiple algorithms for comparison
    communities = {}
    
    # 1. Louvain
    try:
        partition_louvain = community_louvain.best_partition(G)
        communities['louvain'] = partition_louvain
        louvain_modularity = community_louvain.modularity(partition_louvain, G)
    except:
        communities['louvain'] = {n: 0 for n in G.nodes()}
        louvain_modularity = 0
    
    # 2. Label Propagation
    try:
        lpa_communities = list(nx.algorithms.community.label_propagation_communities(G))
        lpa_partition = {n: i for i, comm in enumerate(lpa_communities) for n in comm}
        communities['label_propagation'] = lpa_partition
    except:
        communities['label_propagation'] = {n: 0 for n in G.nodes()}
    
    # 3. Girvan-Newman (first level only for efficiency)
    try:
        comp = nx.community.girvan_newman(G)
        gn_communities = next(comp)
        gn_partition = {n: i for i, comm in enumerate(gn_communities) for n in comm}
        communities['girvan_newman'] = gn_partition
    except:
        communities['girvan_newman'] = {n: 0 for n in G.nodes()}
    
    # 4. Spectral Clustering (using sklearn)
    if G.number_of_nodes() > 1:
        try:
            # Prepare adjacency matrix
            adj_matrix = nx.to_numpy_array(G)
            n_clusters = min(8, G.number_of_nodes() - 1)  # Up to 8 communities
            clustering = SpectralClustering(n_clusters=n_clusters, 
                                          affinity='precomputed',
                                          assign_labels='discretize',
                                          random_state=42)
            labels = clustering.fit_predict(adj_matrix)
            spectral_partition = {node: int(labels[i]) for i, node in enumerate(G.nodes())}
            communities['spectral'] = spectral_partition
        except:
            communities['spectral'] = {n: 0 for n in G.nodes()}
    else:
        communities['spectral'] = {n: 0 for n in G.nodes()}
    
    # 5. Infomap (via cdlib) - try to import cdlib and use it if available
    try:
        infomap = cdlib_algs.infomap(G)
        infomap_partition = {n: c for c, nodes in enumerate(infomap.communities) for n in nodes}
        communities['infomap'] = infomap_partition
    except:
        # Fallback if cdlib not available or infomap fails
        communities['infomap'] = {n: 0 for n in G.nodes()}
    
    # Evaluation of community detection results
    evaluations = {}
    
    # Convert dictionary partitions to list format for cdlib
    def to_coms_list(partition):
        communities_by_id = defaultdict(list)
        for node, com_id in partition.items():
            communities_by_id[com_id].append(node)
        return list(communities_by_id.values())
    
    # Compute evaluation metrics using cdlib
    for algo, partition in communities.items():
        try:
            coms = to_coms_list(partition)
            
            # Cast to cdlib format
            try:
                nodecoms = NodeClustering(
                    communities=coms,
                    graph=G,
                    method_name=algo
                )
                
                # Compute metrics
                evaluations[algo] = {
                    'communities': len(coms),
                    'modularity': evaluation.newman_girvan_modularity(G, nodecoms).score,
                    'internal_density': evaluation.internal_edge_density(G, nodecoms).score,
                    'avg_internal_degree': evaluation.average_internal_degree(G, nodecoms).score,
                    'conductance': evaluation.conductance(G, nodecoms).score,
                }
            except:
                # Fallback if cdlib not available
                evaluations[algo] = {
                    'communities': len(set(partition.values())),
                    'modularity': community_louvain.modularity(partition, G) if algo == 'louvain' else 0.0,
                    'internal_density': 0.0,
                    'avg_internal_degree': 0.0,
                    'conductance': 0.0,
                }
        except:
            evaluations[algo] = {
                'communities': len(set(partition.values())),
                'modularity': 0.0,
                'internal_density': 0.0,
                'avg_internal_degree': 0.0,
                'conductance': 0.0,
            }
    
    # Add community assignments to graph
    for algo, partition in communities.items():
        nx.set_node_attributes(G, partition, algo)
    
    return communities, evaluations, louvain_modularity

# ───────────────────── 7. Visualization Functions ──────────────────────────────   

def plot_network(G, pos=None, metrics=None, community_key='louvain', 
                 node_size_field='eigenvector', title='GitHub Collaboration Network',
                 save_path=None, show_top_nodes=True, top_n=5):
    """Enhanced network visualization with community coloring and node sizing"""
    plt.figure(figsize=(14, 12))
    
    if pos is None:
        # First try ForceAtlas2 for a nicer layout
        try:
            forceatlas2 = ForceAtlas2(
                outboundAttractionDistribution=True,  # Dissuade hubs
                linLogMode=False,  # NOT LinLog mode
                adjustSizes=False,  # Prevent overlap
                edgeWeightInfluence=1.0,
                jitterTolerance=1.0,  # Tolerance
                barnesHutOptimize=True,
                barnesHutTheta=1.2,
                multiThreaded=False,  # NOT for jupyter
                scalingRatio=2.0,
                strongGravityMode=False,
                gravity=1.0,
                verbose=False
            )
            
            # Random initial positions
            positions = {node: [random.random(), random.random()] for node in G.nodes()}
            
            # Run the algorithm for iterations
            pos = forceatlas2.forceatlas2_networkx_layout(G, pos=positions, iterations=1000)
        except Exception as e:
            print(f"ForceAtlas2 layout failed: {e}, falling back to spring layout")
            # Fall back to spring layout
            pos = nx.spring_layout(G, weight='weight', k=0.3, seed=42)
        

        # ─── ensure every node is in pos ─────────────────────────────────
        missing = set(G.nodes()) - set(pos.keys())
        for n in missing:
            # fallback placement for any unlaid-out node
            pos[n] = (random.random(), random.random())
        # ───────────────────────────────────────────────────────────────────
    
    # Get community assignments
    communities = nx.get_node_attributes(G, community_key)
    if not communities:
        # Fallback if specified community attribute doesn't exist
        communities = {n: 0 for n in G.nodes()}
    
    # Map communities to colors
    unique_communities = set(communities.values())
    color_map = {com: PALETTE[i % len(PALETTE)] for i, com in enumerate(unique_communities)}
    node_colors = [color_map[communities[node]] for node in G.nodes()]
    
    # Get node sizes based on the specified metric
    if metrics is None:
        # Default node sizes based on degree if no metrics provided
        node_sizes = [np.log1p(G.degree(n)) * 50 + 20 for n in G.nodes()]
    else:
        # Use the specified metric field
        node_values = [metrics[n][node_size_field] for n in G.nodes()]
        # Scale node sizes logarithmically to handle outliers
        node_sizes = [np.log1p(max(0.001, val)) * 150 + 30 for val in node_values]
    
    # Draw edges with transparency based on weight
    edge_weights = [G[u][v]['weight'] for u, v in G.edges()]
    max_weight = max(edge_weights) if edge_weights else 1
    edge_alphas = [min(0.9, max(0.1, w / max_weight * 0.8)) for w in edge_weights]
    
    # Draw edges with varying width based on weight
    edge_widths = [max(1, w / max_weight * 5) for w in edge_weights]
    
    # Draw the network with nice aesthetics
    nx.draw_networkx_edges(G, pos, alpha=0.3, width=edge_widths, edge_color='gray')
    nodes = nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, alpha=0.9, 
                                  edgecolors='white', linewidths=0.5)
    nodes.set_zorder(2)  # Make sure nodes are on top of edges
    
    # Add labels only for top nodes
    if show_top_nodes:
        # Get top nodes by specified metric
        if metrics is not None:
            top_nodes = sorted(G.nodes(), key=lambda n: metrics[n][node_size_field], reverse=True)[:top_n]
        else:
            top_nodes = sorted(G.nodes(), key=lambda n: G.degree(n), reverse=True)[:top_n]
        
        # Create a subset of the positions dictionary with only top nodes
       
        top_pos = {n: pos[n] for n in top_nodes if n in pos}
        if top_pos:
            # pass the full `pos` layout, then specify `labels=top_pos`
            labels = nx.draw_networkx_labels(
                G,
                pos,
                labels=top_pos,
                font_size=12,
                font_weight='bold'
            )
            for node, text in labels.items():
                text.set_path_effects([pe.withStroke(
                    linewidth=3, foreground='white')])

        for node, text in labels.items():
            text.set_path_effects([pe.withStroke(linewidth=3, foreground='white')])
    
    # Add a title and community legend
    plt.title(title, fontsize=16, fontweight='bold', pad=20)
    
    # Add community legend
    community_patches = []
    for com_id, color in color_map.items():
        community_size = len([n for n in communities if communities[n] == com_id])
        label = f"Community {com_id} ({community_size} members)"
        community_patches.append(mpatches.Patch(color=color, label=label))
    
    plt.legend(handles=community_patches, loc='upper right', title="Communities", 
               fontsize=10, title_fontsize=12, bbox_to_anchor=(1.1, 1.0))
    
    plt.tight_layout()
    plt.axis('off')
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    
    plt.show()
    
    return pos





   
def plot_advanced_metrics(G, metrics, richclub, network_metrics):
    """
    Create a comprehensive dashboard of network metrics with fancy visualizations
    """
    fig = plt.figure(figsize=(20, 16))
    fig.suptitle("Advanced Network Analysis Dashboard", fontsize=20, fontweight='bold', y=0.98)
    
    # Setup subplots grid
    gs = fig.add_gridspec(3, 3)
    
    # 1. Degree distribution with power law fit (log-log plot)
    ax1 = fig.add_subplot(gs[0, 0])
    degrees = [d for n, d in G.degree()]
    degree_counts = Counter(degrees)
    x_values = sorted(degree_counts.keys())
    y_values = [degree_counts[x] for x in x_values]
    
    # Plot the degree distribution
    ax1.loglog(x_values, y_values, 'bo', markersize=5, alpha=0.7, label='Observed')
    
    # Add power law fit if available
    if network_metrics['power_law_alpha'] and network_metrics['power_law_xmin']:
        alpha = network_metrics['power_law_alpha']
        xmin = network_metrics['power_law_xmin']
        # Generate fitted values for plotting
        x_fit = np.logspace(np.log10(xmin), np.log10(max(x_values)), 50)
        y_fit = [x**(-alpha) * (y_values[0] / (x_values[0]**(-alpha))) for x in x_fit]
        ax1.loglog(x_fit, y_fit, 'r-', label=f'Power law fit (α={alpha:.2f})')
    
    ax1.set_title('Degree Distribution', fontweight='bold')
    ax1.set_xlabel('Degree (log scale)')
    ax1.set_ylabel('Frequency (log scale)')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. Rich Club Coefficient Plot
    ax2 = fig.add_subplot(gs[0, 1])
    k_values = sorted(richclub.keys())
    rc_values = [richclub[k] for k in k_values]
    
    ax2.plot(k_values, rc_values, 'go-', linewidth=2)
    ax2.set_title('Rich Club Coefficient', fontweight='bold')
    ax2.set_xlabel('Degree (k)')
    ax2.set_ylabel('φ(k)')
    ax2.grid(True, alpha=0.3)
    
    # 3. Centrality Comparison (Top Contributors)
    ax3 = fig.add_subplot(gs[0, 2])
    top_nodes = sorted(metrics.keys(), key=lambda n: metrics[n]['eigenvector'], reverse=True)[:10]
    centrality_metrics = ['degree', 'betweenness', 'eigenvector', 'closeness']
    
    # Normalize values for comparison
    for metric in centrality_metrics:
        max_value = max(metrics[n][metric] for n in metrics)
        if max_value > 0:  # Avoid division by zero
            for node in metrics:
                metrics[node][f'{metric}_norm'] = metrics[node][metric] / max_value
    
    # Prepare data for plotting
    node_names = [n for n in top_nodes]
    metric_values = {
        'Degree': [metrics[n]['degree_norm'] for n in top_nodes],
        'Betweenness': [metrics[n]['betweenness_norm'] for n in top_nodes],
        'Eigenvector': [metrics[n]['eigenvector_norm'] for n in top_nodes],
        'Closeness': [metrics[n]['closeness_norm'] for n in top_nodes]
    }
    
    # Plot horizontally stacked bar chart for centrality comparison
    bars = ax3.barh(node_names, metric_values['Eigenvector'], label='Eigenvector', color='#3498db')
    prev_values = metric_values['Eigenvector']
    colors = ['#e74c3c', '#2ecc71', '#f39c12']
    for i, metric in enumerate(['Betweenness', 'Closeness', 'Degree']):
        bars = ax3.barh(node_names, metric_values[metric], left=prev_values, label=metric, color=colors[i])
        prev_values = [prev + curr for prev, curr in zip(prev_values, metric_values[metric])]
    
    ax3.set_title('Centrality Comparison (Top 10 Contributors)', fontweight='bold')
    ax3.set_xlabel('Normalized Centrality')
    ax3.legend(loc='lower right')
    ax3.grid(True, axis='x', alpha=0.3)
    
    # 4. Network Structure Metrics Table
    ax4 = fig.add_subplot(gs[1, 0])
    ax4.axis('tight')
    ax4.axis('off')
    
    # Prepare table data
    table_data = [
        ['Nodes', network_metrics['n_nodes']],
        ['Edges', network_metrics['n_edges']],
        ['Density', f"{network_metrics['density']:.4f}"],
        ['Components', network_metrics['components']],
        ['Largest Component', f"{network_metrics['largest_cc_size']} nodes"],
        ['Avg. Clustering', f"{network_metrics['avg_clustering']:.4f}"],
        ['Avg. Path Length', f"{network_metrics['avg_path_length']:.4f}" if network_metrics['avg_path_length'] else "N/A"],
        ['Diameter', f"{network_metrics['diameter']}" if network_metrics['diameter'] else "N/A"],
        ['Degree Assortativity', f"{network_metrics['degree_assortativity']:.4f}" if network_metrics['degree_assortativity'] else "N/A"],
        ['Power Law Exponent', f"{network_metrics['power_law_alpha']:.4f}" if network_metrics['power_law_alpha'] else "N/A"]
    ]
    
    # Create and style the table
    table = ax4.table(cellText=table_data, colLabels=['Metric', 'Value'], 
                     loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1, 1.5)
    table.auto_set_column_width([0, 1])
    
    # Style header row
    for j, cell in enumerate(table._cells[(0, j)] for j in range(2)):
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#2c3e50')
    
    ax4.set_title('Network Structure Metrics', fontweight='bold', pad=20)
    
    # 5. Community Structure Radar Chart
    ax5 = fig.add_subplot(gs[1, 1], polar=True)
    
    # Calculate community-level metrics
    community_key = 'louvain'  # Use louvain as default
    communities = nx.get_node_attributes(G, community_key)
    if not communities:
        communities = {n: 0 for n in G.nodes()}
    
    # Group nodes by community
    community_nodes = defaultdict(list)
    for node, comm in communities.items():
        community_nodes[comm].append(node)
    
    # Calculate metrics for each community
    community_metrics = {}
    for comm, nodes in community_nodes.items():
        subgraph = G.subgraph(nodes)
        community_metrics[comm] = {
            'size': len(nodes),
            'density': nx.density(subgraph) if len(nodes) > 1 else 0,
            'avg_clustering': nx.average_clustering(subgraph) if len(nodes) > 1 else 0,
            'avg_degree': sum(dict(subgraph.degree()).values()) / len(nodes) if len(nodes) > 0 else 0,
            # Influence: percentage of total eigenvector centrality
            'influence': sum(metrics[n]['eigenvector'] for n in nodes if n in metrics) / 
                        sum(metrics[n]['eigenvector'] for n in metrics) if metrics else 0
        }
    
    # Select top communities by size for radar chart
    top_communities = sorted(community_metrics.keys(), key=lambda c: community_metrics[c]['size'], reverse=True)[:5]
    
    # Set up radar chart
    angles = np.linspace(0, 2*np.pi, 5, endpoint=False).tolist()
    angles += angles[:1]  # Close the loop
    
    metrics_names = ['Size', 'Density', 'Clustering', 'Avg Degree', 'Influence']
    
    # Plot each community as a polygon on the radar chart
    for i, comm in enumerate(top_communities):
        values = [
            community_metrics[comm]['size'] / max(cm['size'] for cm in community_metrics.values()),
            community_metrics[comm]['density'],
            community_metrics[comm]['avg_clustering'],
            community_metrics[comm]['avg_degree'] / max(cm['avg_degree'] for cm in community_metrics.values()),
            community_metrics[comm]['influence']
        ]
        values += values[:1]  # Close the loop
        
        ax5.plot(angles, values, 'o-', linewidth=2, label=f'Community {comm}', color=PALETTE[i % len(PALETTE)])
        ax5.fill(angles, values, alpha=0.1, color=PALETTE[i % len(PALETTE)])
    
    # Set radar chart properties
    ax5.set_xticks(angles[:-1])
    ax5.set_xticklabels(metrics_names)
    ax5.set_yticks([0.2, 0.4, 0.6, 0.8])
    ax5.set_yticklabels(['0.2', '0.4', '0.6', '0.8'], color='gray')
    ax5.set_title('Community Structure Analysis', fontweight='bold')
    ax5.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
    
    # 6. Structural Roles Distribution
    ax6 = fig.add_subplot(gs[1, 2])
    
    # Get role assignment from metrics
    role_counts = Counter([metrics[n]['structural_role'] for n in metrics])
    
    # Prepare data for pie chart
    role_labels = list(role_counts.keys())
    role_sizes = list(role_counts.values())
    
    # Custom color map for roles
    role_colors = PALETTE[:len(role_labels)]
    
    # Create pie chart
    wedges, texts, autotexts = ax6.pie(
        role_sizes, 
        labels=None,
        autopct='%1.1f%%',
        startangle=90,
        colors=role_colors,
        wedgeprops={'width': 0.6, 'edgecolor': 'w'}
    )
    
    # Customize pie chart text
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontsize(9)
        autotext.set_weight('bold')
    
    # Add legend
    ax6.legend(wedges, [f"{label} ({size})" for label, size in zip(role_labels, role_sizes)],
              title="Structural Roles",
              loc="center left",
              bbox_to_anchor=(1, 0, 0.5, 1))
    
    ax6.set_title('Structural Role Distribution', fontweight='bold')
    
    # 7. Top Influencers Bar Chart (Bottom row, spanning 2 columns)
    ax7 = fig.add_subplot(gs[2, :2])
    
    # Get top influencers
    top_influencers = sorted([(n, metrics[n]['influencer_score']) for n in metrics], 
                           key=lambda x: x[1], reverse=True)[:10]
    
    names = [name for name, _ in top_influencers]
    scores = [score for _, score in top_influencers]
    
    # Create horizontal bar chart
    bars = ax7.barh(range(len(names)), scores, color=PALETTE[3], alpha=0.8, 
                  edgecolor='white', linewidth=1, height=0.6)
    
    # Add values to the end of the bars
    for i, (bar, score) in enumerate(zip(bars, scores)):
        ax7.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2, 
                f'{score:.3f}', va='center')
    
    ax7.set_yticks(range(len(names)))
    ax7.set_yticklabels(names)
    ax7.set_title('Top 10 Influencers by Composite Score', fontweight='bold')
    ax7.set_xlabel('Influencer Score')
    ax7.invert_yaxis()  # Make the top influencer appear at the top
    ax7.grid(True, axis='x', alpha=0.3)
    
    # 8. Governance Recommendations (Bottom-right)
    ax8 = fig.add_subplot(gs[2, 2])
    ax8.axis('off')
    
    # Generate recommendations based on network analysis
    recommendations = [
        "1. Encourage bridge nodes to mentor newcomers",
        "2. Promote cross-community collaboration",
        "3. Recognize hub contributors with special roles",
        "4. Create focused working groups around communities",
        "5. Identify and nurture potential new leaders"
    ]
    
    # Add recommendations as text
    ax8.text(0.5, 0.95, "Project Governance Recommendations", fontsize=14, fontweight='bold', 
            ha='center', va='top')
    
    y_pos = 0.85
    for rec in recommendations:
        ax8.text(0.1, y_pos, rec, fontsize=11, 
                va='top', wrap=True, 
                bbox=dict(boxstyle="round,pad=0.5", fc="#f8f9fa", ec="lightgray", alpha=0.8))
        y_pos -= 0.15
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for suptitle
    plt.show()

def create_temporal_animation(edges_df, time_windows=TIME_WINDOWS, interval=50, save_path=None):
    """
    Create an animated visualization of network evolution over time
    """
    if edges_df is None or edges_df.empty:
        print("No temporal data available for animation")
        return None
    
    # Ensure timestamp column exists
    if 'timestamp' not in edges_df.columns:
        print("Timestamp column missing from edges dataframe")
        return None
    
    # Convert timestamps to datetime if needed
    if not pd.api.types.is_datetime64_any_dtype(edges_df['timestamp']):
        edges_df['timestamp'] = pd.to_datetime(edges_df['timestamp'])
    
    # Sort edges by timestamp
    edges_df = edges_df.sort_values('timestamp')
    
    # Calculate time intervals
    start_time = edges_df['timestamp'].min()
    end_time = edges_df['timestamp'].max()
    time_delta = (end_time - start_time) / time_windows
    
    # Create figure
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Initialize empty graph
    G_cum = nx.Graph()
    pos = None
    
    # Function to update the graph for each frame
    def update(frame):
        nonlocal G_cum, pos
        
        # Clear previous frame
        ax.clear()
        
        # Calculate cutoff time for this frame
        cutoff_time = start_time + (frame + 1) * time_delta
        
        # Get edges up to this point in time
        current_edges = edges_df[edges_df['timestamp'] <= cutoff_time]
        
        # Add edges to cumulative graph
        for _, row in current_edges.iterrows():
            if row['source'] not in G_cum:
                G_cum.add_node(row['source'])
            if row['target'] not in G_cum:
                G_cum.add_node(row['target'])
            
            if G_cum.has_edge(row['source'], row['target']):
                G_cum[row['source']][row['target']]['weight'] += 1
            else:
                G_cum.add_edge(row['source'], row['target'], weight=1)
        
        # Calculate layout if not already done (only once for stability)
        if pos is None and G_cum.number_of_nodes() > 0:
            pos = nx.spring_layout(G_cum, k=0.3, seed=42)
        
        # If still no positions (e.g. no nodes yet), create empty positions
        if pos is None:
            pos = {}
        
        # Add any new nodes to layout
        for node in G_cum.nodes():
            if node not in pos:
                # Try to place near neighbors if possible
                neighbors = list(G_cum.neighbors(node))
                if neighbors:
                    neighbor_pos = [pos[neigh] for neigh in neighbors if neigh in pos]
                    if neighbor_pos:
                        # Place near the average position of neighbors
                        avg_x = sum(p[0] for p in neighbor_pos) / len(neighbor_pos)
                        avg_y = sum(p[1] for p in neighbor_pos) / len(neighbor_pos)
                        # Add small random offset
                        pos[node] = (avg_x + random.uniform(-0.1, 0.1), 
                                    avg_y + random.uniform(-0.1, 0.1))
                    else:
                        pos[node] = (random.uniform(-1, 1), random.uniform(-1, 1))
                else:
                    pos[node] = (random.uniform(-1, 1), random.uniform(-1, 1))
        
        # Draw the network
        if G_cum.number_of_nodes() > 0:
            # Get edge weights for line thickness
            edge_weights = [G_cum[u][v]['weight'] for u, v in G_cum.edges()]
            max_weight = max(edge_weights) if edge_weights else 1
            edge_widths = [max(1, w/max_weight * 5) for w in edge_weights]
            
            # Calculate node sizes based on degree
            node_sizes = [20 + 10 * G_cum.degree(n) for n in G_cum.nodes()]
            
            # Use degree for coloring nodes
            node_degrees = [G_cum.degree(n) for n in G_cum.nodes()]
            vmin = min(node_degrees)
            vmax = max(node_degrees) if node_degrees else 1
            
            # Draw the network
            nx.draw_networkx_edges(G_cum, pos, alpha=0.4, width=edge_widths, edge_color='gray')
            nodes = nx.draw_networkx_nodes(G_cum, pos, 
                                          node_size=node_sizes,
                                          node_color=node_degrees, 
                                          cmap='viridis',
                                          vmin=vmin, vmax=vmax,
                                          alpha=0.8)
            
            # Add timestamp as title
            timestamp_str = cutoff_time.strftime('%Y-%m-%d')
            ax.set_title(f"Network Evolution: {timestamp_str}\nNodes: {G_cum.number_of_nodes()}, Edges: {G_cum.number_of_edges()}", 
                        fontweight='bold')
            
            # Add colorbar
            if 'cbar' not in locals():
                sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=vmin, vmax=vmax))
                sm.set_array([])
                plt.colorbar(sm, ax=ax, label='Node Degree')
        
        ax.set_axis_off()
        return fig,
    
    # Create animation
    anim = animation.FuncAnimation(fig, update, frames=time_windows, interval=interval, blit=False)
    
    # Save animation if path provided
    if save_path:
        anim.save(save_path, writer='pillow', fps=2, dpi=100)
    
    plt.close()
    
    return anim

def create_interactive_plots(G, metrics, temporal_metrics=None):
    """
    Create interactive visualizations using Plotly
    """
    # 1. Create an interactive network visualization
    # Prepare node data
    node_x = []
    node_y = []
    node_text = []
    node_size = []
    node_color = []
    
    # Get positions for nodes (use spring layout)
    pos = nx.spring_layout(G, seed=42)
    
    # Get community assignments if available
    communities = nx.get_node_attributes(G, 'louvain')
    if not communities:
        communities = {n: 0 for n in G.nodes()}
    
    for node in G.nodes():
        x, y = pos[node]
        node_x.append(x)
        node_y.append(y)
        
        # Node info for hover
        degree = G.degree(node)
        eigc = metrics[node]['eigenvector'] if node in metrics else 0
        betw = metrics[node]['betweenness'] if node in metrics else 0
        community = communities[node] if node in communities else 0
        
        node_text.append(f"User: {node}<br>Degree: {degree}<br>Eigenvector: {eigc:.4f}<br>"
                         f"Betweenness: {betw:.4f}<br>Community: {community}")
        
        # Node size based on eigenvector centrality
        node_size.append(metrics[node]['eigenvector'] * 100 + 10 if node in metrics else 10)
        
        # Node color based on community
        node_color.append(communities[node] if node in communities else 0)
    
    # Create edges
    edge_x = []
    edge_y = []
    
    for u, v in G.edges():
        x0, y0 = pos[u]
        x1, y1 = pos[v]
        edge_x.extend([x0, x1, None])
        edge_y.extend([y0, y1, None])
    
    # Create edge trace
    edge_trace = go.Scatter(
        x=edge_x, y=edge_y,
        line=dict(width=0.5, color='#888'),
        hoverinfo='none',
        mode='lines',
        opacity=0.7
    )
    
    # Create node trace
    node_trace = go.Scatter(
        x=node_x, y=node_y,
        mode='markers',
        hoverinfo='text',
        text=node_text,
        # marker=dict(
        #     showscale=True,
        #     colorscale='Viridis',
        #     color=node_color,
        #     size=node_size,
        #     colorbar=dict(
        #         thickness=15,
        #         title='Community',
        #         xanchor='left',
        #         titleside='right'
        #     ),
        #     line=dict(width=2, color='white')
        # )

        marker=dict(
            showscale=True,
            colorscale='Viridis',
            color=node_color,
            size=node_size,
            colorbar=dict(
                thickness=15,
                xanchor='left',
                title=dict(               # ▶️ move into a dict
                    text='Community',     # ▶️ the bar’s title text
                    side='right'          # ▶️ replaces old `titleside` 
                )
            ),
            line=dict(width=2, color='white')
        )
    )
    
    # Create the figure
    fig_network = go.Figure(
        data=[edge_trace, node_trace],
        layout=go.Layout(
            title=dict(                       # ✅  new API
                text='Interactive GitHub Collaboration Network',
                font=dict(size=16)
            ),

            showlegend=False,
            hovermode='closest',
            margin=dict(b=20, l=5, r=5, t=40),
            xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
            yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
            plot_bgcolor='rgba(240,240,240,0.8)',
            paper_bgcolor='white',
            autosize=True,
            height=700
        )
    )
    
    # 2. Create radar chart for top contributors (multi-metric comparison)
    # Get top contributors by influence score
    top_contributors = sorted(metrics.keys(), key=lambda n: metrics[n]['influencer_score'], reverse=True)[:8]
    
    # Prepare data for radar chart
    categories = ['Degree', 'Betweenness', 'Eigenvector', 'Closeness', 'Clustering']
    
    # Normalize values for radar chart
    max_values = {
        'Degree': max(metrics[n]['degree'] for n in metrics) if metrics else 1,
        'Betweenness': max(metrics[n]['betweenness'] for n in metrics) if metrics else 1,
        'Eigenvector': max(metrics[n]['eigenvector'] for n in metrics) if metrics else 1,
        'Closeness': max(metrics[n]['closeness'] for n in metrics) if metrics else 1,
        'Clustering': max(metrics[n]['clustering'] for n in metrics) if metrics else 1
    }
    
    # Create radar chart
    fig_radar = go.Figure()
    
    for i, contributor in enumerate(top_contributors):
        values = [
            metrics[contributor]['degree'] / max_values['Degree'],
            metrics[contributor]['betweenness'] / max_values['Betweenness'],
            metrics[contributor]['eigenvector'] / max_values['Eigenvector'],
            metrics[contributor]['closeness'] / max_values['Closeness'],
            metrics[contributor]['clustering'] / max_values['Clustering']
        ]
        # Add the first value again to close the polygon
        categories_closed = categories + [categories[0]]
        values_closed = values + [values[0]]
        
        fig_radar.add_trace(go.Scatterpolar(
            r=values_closed,
            theta=categories_closed,
            fill='toself',
            name=contributor,
            line_color=px.colors.qualitative.Plotly[i % len(px.colors.qualitative.Plotly)]
        ))
    
    fig_radar.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0, 1.1]
            )
        ),
        title='Top Contributors Influence Profile',
        showlegend=True,
        height=600
    )
    
    # 3. Create temporal evolution chart if temporal data is available
    if temporal_metrics is not None:
        # Create a subplot with multiple time series
        fig_temporal = make_subplots(
            rows=2, cols=2,
            subplot_titles=(
                'Nodes Over Time', 
                'Edges Over Time',
                'Density & Clustering Over Time',
                'Communities & Modularity Over Time'
            )
        )
        
        # Add time series for nodes
        fig_temporal.add_trace(
            go.Scatter(
                x=temporal_metrics['time'], 
                y=temporal_metrics['nodes'],
                mode='lines+markers',
                name='Nodes',
                line=dict(color=PALETTE[0], width=3)
            ),
            row=1, col=1
        )
        
        # Add time series for edges
        fig_temporal.add_trace(
            go.Scatter(
                x=temporal_metrics['time'], 
                y=temporal_metrics['edges'],
                mode='lines+markers',
                name='Edges',
                line=dict(color=PALETTE[1], width=3)
            ),
            row=1, col=2
        )
        
        # Add time series for density and clustering
        fig_temporal.add_trace(
            go.Scatter(
                x=temporal_metrics['time'], 
                y=temporal_metrics['density'],
                mode='lines+markers',
                name='Density',
                line=dict(color=PALETTE[2], width=3)
            ),
            row=2, col=1
        )
        
        fig_temporal.add_trace(
            go.Scatter(
                x=temporal_metrics['time'], 
                y=temporal_metrics['avg_clustering'],
                mode='lines+markers',
                name='Avg Clustering',
                line=dict(color=PALETTE[3], width=3, dash='dash')
            ),
            row=2, col=1
        )
        
        # Add time series for communities and modularity
        fig_temporal.add_trace(
            go.Scatter(
                x=temporal_metrics['time'], 
                y=temporal_metrics['communities'],
                mode='lines+markers',
                name='Communities',
                line=dict(color=PALETTE[4], width=3)
            ),
            row=2, col=2
        )
        
        fig_temporal.add_trace(
            go.Scatter(
                x=temporal_metrics['time'], 
                y=temporal_metrics['modularity'],
                mode='lines+markers',
                name='Modularity',
                line=dict(color=PALETTE[5], width=3, dash='dash')
            ),
            row=2, col=2
        )
        
        # Update layout
        fig_temporal.update_layout(
            title_text='Network Metrics Evolution Over Time',
            height=800,
            width=1000,
            showlegend=True
        )
        
        return fig_network, fig_radar, fig_temporal
    
    return fig_network, fig_radar

def generate_network_summary_report(G, metrics, network_metrics, communities_eval, richclub, temporal_windows=None):
    """
    Generate a comprehensive network analysis report with recommendations for project governance
    """
    # Get basic metrics
    n_nodes = G.number_of_nodes()
    n_edges = G.number_of_edges()
    avg_degree = 2 * n_edges / n_nodes if n_nodes > 0 else 0
    
    # Find top contributors by different metrics
    top_degree = sorted(G.nodes(), key=lambda n: metrics[n]['degree'], reverse=True)[:5]
    top_betweenness = sorted(G.nodes(), key=lambda n: metrics[n]['betweenness'], reverse=True)[:5]
    top_eigenvector = sorted(G.nodes(), key=lambda n: metrics[n]['eigenvector'], reverse=True)[:5]
    top_influencers = sorted(G.nodes(), key=lambda n: metrics[n]['influencer_score'], reverse=True)[:5]
    top_bridges = sorted(G.nodes(), key=lambda n: metrics[n]['bridge_score'], reverse=True)[:5]
    
    # Get largest connected component
    components = list(nx.connected_components(G))
    largest_cc = max(components, key=len) if components else set()

    
    alpha_fmt = f"{network_metrics['power_law_alpha']:.2f}" if network_metrics.get('power_law_alpha') else "N/A"
    clustering_fmt = f"{network_metrics['avg_clustering']:.3f}"



    # Create the report
    report = f"""
    # GitHub Collaboration Network Analysis Report
    
    ## Executive Summary
    
    Analysis of collaboration across {n_nodes} contributors and {n_edges} interactions reveals a 
    {network_metrics['components']}-component network with {communities_eval['louvain']['communities']} distinct communities.
    
    The network exhibits a **{'scale-free' if network_metrics.get('power_law_alpha',0) and 2 < network_metrics['power_law_alpha'] < 3 else 'non-scale-free'}** structure
    with power law exponent α={alpha_fmt}.
    
    Contributors show an average clustering coefficient of {network_metrics['avg_clustering']:.3f}, indicating
    {'high' if network_metrics['avg_clustering'] > 0.3 else 'moderate' if network_metrics['avg_clustering'] > 0.1 else 'low'} levels of local clustering.
    
    ## Key Influencers
    
    **Hub Contributors** (high degree centrality):
    {', '.join([f"{node} ({metrics[node]['degree']})" for node in top_degree])}
    
    **Bridge Contributors** (high betweenness centrality):
    {', '.join([f"{node} ({metrics[node]['betweenness']:.2f})" for node in top_betweenness])}
    
    **Core Contributors** (high eigenvector centrality):
    {', '.join([f"{node} ({metrics[node]['eigenvector']:.2f})" for node in top_eigenvector])}
    
    **Overall Influencers** (composite score):
    {', '.join([f"{node} ({metrics[node]['influencer_score']:.2f})" for node in top_influencers])}
    
    ## Community Structure
    
    The repository collaboration network contains {communities_eval['louvain']['communities']} distinct communities,
    with modularity of {communities_eval['louvain']['modularity']:.3f}, indicating
    {'strong' if communities_eval['louvain']['modularity'] > 0.5 else 'moderate' if communities_eval['louvain']['modularity'] > 0.3 else 'weak'} 
    community structure.
    
    ## Rich Club Analysis
    
    {'A significant rich club effect is present, with high-degree nodes showing preferential attachment to each other.' 
    if any(v > 1.5 for k, v in richclub.items() if k > 3) else
    'No significant rich club effect detected, suggesting an egalitarian contribution structure.'}
    
    ## Structural Roles
    
    The network contains diverse structural roles:
    - **Hubs**: {sum(1 for n in metrics if metrics[n]['structural_role'] == 'Hub')} contributors
    - **Bridges**: {sum(1 for n in metrics if metrics[n]['structural_role'] == 'Bridge')} contributors  
    - **Peripheral**: {sum(1 for n in metrics if metrics[n]['structural_role'] == 'Peripheral')} contributors
    - **Core**: {sum(1 for n in metrics if metrics[n]['structural_role'] == 'Core')} contributors
    - **Regular**: {sum(1 for n in metrics if metrics[n]['structural_role'] == 'Regular')} contributors
    
    ## Recommendations for Project Governance
    
    1. **Community Integration**:
       - Create cross-community initiatives led by bridge nodes ({', '.join(top_bridges[:2])})
       - Establish regular knowledge-sharing sessions between community representatives
    
    2. **Contributor Development**:
       - Implement mentorship program pairing peripheral nodes with bridge nodes
       - Create documentation guidelines led by hub contributors ({', '.join(top_degree[:2])})
    
    3. **Sustainability Planning**:
       - Identify potential successors for core contributors
       - Distribute critical knowledge across multiple contributors
    
    4. **Community Growth**:
       - Target recruitment to strengthen smaller communities
       - Create "welcome teams" with representatives from each community
    
    5. **Collaboration Enhancement**:
       - Implement rotation system for cross-community review
       - Establish formal knowledge transfer protocols
    """
    
    if temporal_windows:
        # Add temporal analysis section if available
        report += f"""
    ## Temporal Analysis
    
    The network has shown {'growth' if temporal_windows[-1]['n_nodes'] > temporal_windows[0]['n_nodes'] else 'decline'} 
    in active contributors over the analyzed period.
    
    Community structure has {'strengthened' if temporal_windows[-1]['modularity'] > temporal_windows[0]['modularity'] else 'weakened'} 
    over time, with modularity {'increasing' if temporal_windows[-1]['modularity'] > temporal_windows[0]['modularity'] else 'decreasing'} 
    from {temporal_windows[0]['modularity']:.3f} to {temporal_windows[-1]['modularity']:.3f}.
    """
    
    return report

def plot_community_evaluation(communities_eval):
    """
    Visualize the comparison of different community detection algorithms
    """
    # Extract algorithm names and metrics
    algorithms = list(communities_eval.keys())
    metrics = [
        'communities', 'modularity', 'internal_density', 
        'avg_internal_degree', 'conductance'
    ]
    
    # Create figure with subplots
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Community Detection Algorithm Comparison', fontsize=18, fontweight='bold')
    
    # Flatten axes for easier iteration
    axes = axes.flatten()
    
    # 1. Number of communities
    ax = axes[0]
    community_counts = [communities_eval[algo]['communities'] for algo in algorithms]
    bars = ax.bar(algorithms, community_counts, color=PALETTE[:len(algorithms)])
    ax.set_title('Number of Communities Detected', fontweight='bold')
    ax.set_ylabel('Count')
    ax.set_xticklabels(algorithms, rotation=45)
    
    # Add value labels on bars
    for bar, count in zip(bars, community_counts):
        ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.1, 
                str(count), ha='center', va='bottom')
    
    # 2. Modularity comparison
    ax = axes[1]
    modularity_values = [communities_eval[algo]['modularity'] for algo in algorithms]
    bars = ax.bar(algorithms, modularity_values, color=PALETTE[:len(algorithms)])
    ax.set_title('Modularity (Higher is Better)', fontweight='bold')
    ax.set_ylabel('Modularity Score')
    ax.set_xticklabels(algorithms, rotation=45)
    
    # Add value labels on bars
    for bar, val in zip(bars, modularity_values):
        ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.01, 
                f"{val:.3f}", ha='center', va='bottom')
    
    # 3. Internal density and average internal degree
    ax = axes[2]
    
    # Setup x positions for grouped bars
    x = np.arange(len(algorithms))
    width = 0.35
    
    # Create grouped bar chart
    density_values = [communities_eval[algo]['internal_density'] for algo in algorithms]
    avg_degree_values = [communities_eval[algo]['avg_internal_degree'] for algo in algorithms]
    
    # Normalize avg_degree for comparison
    max_degree = max(avg_degree_values)
    norm_degree_values = [d / max_degree for d in avg_degree_values] if max_degree > 0 else avg_degree_values
    
    bars1 = ax.bar(x - width/2, density_values, width, label='Internal Density', color='#3498db')
    bars2 = ax.bar(x + width/2, norm_degree_values, width, label='Normalized Avg Internal Degree', color='#e74c3c')
    
    ax.set_title('Community Cohesion Metrics', fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(algorithms, rotation=45)
    ax.legend()
    
    # Add value labels on bars
    for bar, val in zip(bars1, density_values):
        ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.01, 
                f"{val:.3f}", ha='center', va='bottom', fontsize=9)
    
    for bar, val in zip(bars2, norm_degree_values):
        ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.01, 
                f"{val:.3f}", ha='center', va='bottom', fontsize=9)
    
    # 4. Conductance (lower is better)
    ax = axes[3]
    conductance_values = [communities_eval[algo]['conductance'] for algo in algorithms]
    bars = ax.bar(algorithms, conductance_values, color=PALETTE[:len(algorithms)])
    ax.set_title('Conductance (Lower is Better)', fontweight='bold')
    ax.set_ylabel('Conductance Score')
    ax.set_xticklabels(algorithms, rotation=45)
    
    # Add value labels on bars
    for bar, val in zip(bars, conductance_values):
        ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.01, 
                f"{val:.3f}", ha='center', va='bottom')
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust for the suptitle
    plt.show()

# ───────────────────── 8. Main Analysis Function ──────────────────────────────
def analyze_github_network(repos=None, save_path=None, verbose=True):
    """
    Main function to perform comprehensive network analysis on GitHub repositories
    """
    if repos is None:
        repos = [
            "pytorch/pytorch", 
            "huggingface/transformers"
        ]
    
    all_edges = []
    all_metadata = {}
    
    # Fetch data
    for repo in repos:
        if verbose:
            print(f"Fetching data for {repo}...")
        
        try:
            edges, metadata = fetch_user_edges_api(repo, max_pages=1, with_timestamps=True)
            all_edges.extend(edges)
            all_metadata.update(metadata)
            
            if verbose:
                print(f"  → {len(edges)} edges collected from {repo}")
        except Exception as e:
            print(f"Error fetching data for {repo}: {e}")
    
    # Build the graph
    G, G_temporal, edges_df = build_graph(all_edges, min_interactions=1, with_timestamps=True)
    
    if verbose:
        print(f"Graph: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")
    
    # Compute metrics
    node_metrics, richclub_data, network_metrics = novel_metrics(G)
    
    # Detect and evaluate communities
    communities, community_evaluations, louvain_modularity = community_scan(G)
    
    if verbose:
        print(f"Louvain communities: {community_evaluations['louvain']['communities']}  |  "
              f"Modularity: {louvain_modularity:.4f}")
    
    # Show top contributors by eigenvector centrality
    if verbose:
        print("Top 5 by eigenvector centrality:")
        top_eig = sorted(G.nodes(), key=lambda n: node_metrics[n]['eigenvector'], reverse=True)[:5]
        for node in top_eig:
            print(f"  • {node:<20} eig={node_metrics[node]['eigenvector']:.3f}  k={node_metrics[node]['degree']}")
        
        print(f"\nAvg clustering coef: {nx.average_clustering(G)}")
        # Get largest connected component for path length calculation
        largest_cc = max(nx.connected_components(G), key=len)
        G_lcc = G.subgraph(largest_cc)
        if nx.is_connected(G_lcc) and G_lcc.number_of_nodes() > 0:
            print(f"Avg shortest path (LCC): {nx.average_shortest_path_length(G_lcc)}")
    
    # Perform temporal analysis if data available
    temporal_windows = None
    if edges_df is not None and not edges_df.empty:
        temporal_windows, temporal_metrics = temporal_network_evolution(edges_df)
    
    # Create visualizations
    if save_path:
        # Save static network visualization
        pos = plot_network(G, metrics=node_metrics, save_path=f"{save_path}/network.png")
        
        # Create and save advanced metrics dashboard
        plot_advanced_metrics(G, node_metrics, richclub_data, network_metrics)
        plt.savefig(f"{save_path}/metrics_dashboard.png", dpi=300, bbox_inches='tight')
        
        # Create and save community evaluation comparison
        plot_community_evaluation(community_evaluations)
        plt.savefig(f"{save_path}/community_comparison.png", dpi=300, bbox_inches='tight')
        
        # Create and save temporal animation if data available
        if edges_df is not None and not edges_df.empty:
            create_temporal_animation(edges_df, save_path=f"{save_path}/temporal_evolution.gif")
        
        # Create interactive plots
        if edges_df is not None and not edges_df.empty:
            fig_network, fig_radar, fig_temporal = create_interactive_plots(G, node_metrics, temporal_metrics)
            fig_network.write_html(f"{save_path}/interactive_network.html")
            fig_radar.write_html(f"{save_path}/contributor_profiles.html")
            fig_temporal.write_html(f"{save_path}/temporal_metrics.html")
        else:
            fig_network, fig_radar = create_interactive_plots(G, node_metrics)
            fig_network.write_html(f"{save_path}/interactive_network.html")
            fig_radar.write_html(f"{save_path}/contributor_profiles.html")
    else:
        # Display visualizations
        pos = plot_network(G, metrics=node_metrics)
        plot_advanced_metrics(G, node_metrics, richclub_data, network_metrics)
        plot_community_evaluation(community_evaluations)
        
        if edges_df is not None and not edges_df.empty:
            create_temporal_animation(edges_df)
            fig_network, fig_radar, fig_temporal = create_interactive_plots(G, node_metrics, temporal_metrics)
            fig_network.show()
            fig_radar.show()
            fig_temporal.show()
        else:
            fig_network, fig_radar = create_interactive_plots(G, node_metrics)
            fig_network.show()
            fig_radar.show()
    
    # Generate summary report
    report = generate_network_summary_report(G, node_metrics, network_metrics, 
                                           community_evaluations, richclub_data, 
                                           temporal_windows)
    
    if save_path:
        with open(f"{save_path}/network_analysis_report.md", "w") as f:
            f.write(report)
    
    if verbose:
        print("\n" + "="*80)
        print("Network Analysis Summary:")
        print("="*80)
        print(report)
    
    return {
        'graph': G,
        'temporal_graph': G_temporal,
        'edges_df': edges_df,
        'metrics': node_metrics,
        'network_metrics': network_metrics,
        'communities': communities,
        'community_evaluations': community_evaluations,
        'richclub': richclub_data,
        'temporal_windows': temporal_windows,
        'report': report
    }



# ───────────────────── 9. Command-Line Interface ──────────────────────────────
if __name__ == "__main__":
    import argparse
    import os
    
    parser = argparse.ArgumentParser(description="GitHub Collaboration Network Analysis")
    parser.add_argument("--repos", nargs="+", 
                       help="GitHub repositories to analyze (format: owner/repo)")
    parser.add_argument("--token", 
                       help="GitHub API token (alternatively set the TOKEN environment variable)")
    parser.add_argument("--output", "-o", 
                       help="Output directory for visualizations and report")
    parser.add_argument("--quiet", "-q", action="store_true", 
                       help="Suppress verbose output")
    
    # args = parser.parse_args()
    args, unknown = parser.parse_known_args()

    # Set GitHub token
    if args.token:
        TOKEN = args.token
    elif "GITHUB_TOKEN" in os.environ:
        TOKEN = os.environ["GITHUB_TOKEN"]
    
    # Create output directory if specified
    if args.output:
        os.makedirs(args.output, exist_ok=True)
    
    # Run analysis
    # analyze_github_network(
    #     repos=args.repos,
    #     save_path=args.output,
    #     verbose=not args.quiet
    # )
```

```python
import json, datetime

"""
GitHub Network Analysis - Additional Implementation

This script implements the missing parts of the GitHub network analysis project:
1. Community Detection and Influence Analysis
2. Temporal Evolution Study 
3. Practical Recommendations & Project Governance Toolkit

Note: Assumes all functions from the provided code are already implemented.
"""

# ───────────────────── 1. Community Detection and Influence Analysis ─────────────────────

def analyze_communities_and_influence(G, metrics=None, communities=None):
    """
    Perform detailed analysis of community structure and influencer patterns
    
    Parameters:
    - G: NetworkX graph object
    - metrics: Pre-computed node metrics dictionary
    - communities: Pre-computed communities
    
    Returns:
    - Dictionary with community analysis and influence rankings
    """
    results = {}
    
    if not metrics or not communities:
        # Generate metrics and communities if not provided
        node_metrics, richclub_data, network_metrics = novel_metrics(G)
        communities, community_evaluations, _ = community_scan(G)
        metrics = node_metrics
    else:
        node_metrics = metrics
    
    # Compare community detection algorithms
    alg_comparison = {}
    community_types = ['louvain', 'leiden', 'infomap', 'label_prop']
    
    # Extract community sizes for each algorithm
    for alg in community_types:
        if alg in communities:
            comm_sizes = [len(c) for c in communities[alg]]
            alg_comparison[alg] = {
                'count': len(comm_sizes),
                'sizes': comm_sizes,
                'max_size': max(comm_sizes),
                'min_size': min(comm_sizes),
                'avg_size': sum(comm_sizes)/len(comm_sizes)
            }
    
    results['algorithm_comparison'] = alg_comparison
    
    # Calculate inter-community connections
    inter_community_edges = {}
    if 'louvain' in communities:  # Use Louvain as baseline
        node_to_comm = {}
        for i, comm in enumerate(communities['louvain']):
            for node in comm:
                node_to_comm[node] = i
                
        comm_connections = defaultdict(int)
        for u, v in G.edges():
            if u in node_to_comm and v in node_to_comm:
                c1, c2 = node_to_comm[u], node_to_comm[v]
                if c1 != c2:
                    key = tuple(sorted([c1, c2]))
                    comm_connections[key] += 1
        
        inter_community_edges = dict(comm_connections)
        results['inter_community_connections'] = inter_community_edges
    
    # Identify top influencers using different centrality measures
    # We'll create rankings based on different metrics
    top_contributors = {}
    
    centrality_measures = [
        'degree', 'betweenness', 'eigenvector', 'pagerank', 
        'influencer_score', 'bridge_score'
    ]
    
    for measure in centrality_measures:
        if all(measure in metrics[node] for node in G.nodes()):
            ranked_nodes = sorted(G.nodes(), key=lambda n: metrics[n].get(measure, 0), reverse=True)
            top_contributors[measure] = [(node, metrics[node].get(measure, 0)) for node in ranked_nodes[:10]]
    
    results['top_contributors'] = top_contributors
    
    # Analyze contribution patterns within each community
    community_profiles = {}
    
    if 'louvain' in communities:
        for i, comm in enumerate(communities['louvain']):
            # Skip extremely small communities
            if len(comm) < 3:
                continue
                
            # Create subgraph for this community
            subgraph = G.subgraph(comm)
            
            # Compute community-level metrics
            density = nx.density(subgraph)
            
            # Find top nodes within this community
            comm_nodes = list(subgraph.nodes())
            top_node = max(comm_nodes, key=lambda n: metrics[n].get('degree', 0))
            
            # Calculate average metrics within community
            avg_degree = sum(metrics[n].get('degree', 0) for n in comm_nodes) / len(comm_nodes)
            
            # Find bridges to other communities
            bridges = []
            for node in comm_nodes:
                neighbors = set(G.neighbors(node))
                external = [n for n in neighbors if n not in comm]
                if external:
                    bridges.append((node, len(external)))
            
            # Sort bridges by number of external connections
            bridges.sort(key=lambda x: x[1], reverse=True)
            
            community_profiles[i] = {
                'size': len(comm),
                'density': density,
                'top_node': top_node,
                'avg_degree': avg_degree,
                'bridges': bridges[:3],  # Top 3 bridges
                'cohesion': nx.average_clustering(subgraph) if len(comm) > 2 else 0
            }
    
    results['community_profiles'] = community_profiles
    
    # Analyze influencer roles across communities
    cross_community_influencers = []
    
    if 'louvain' in communities:
        # Create a mapping of nodes to their communities
        node_communities = {}
        for i, comm in enumerate(communities['louvain']):
            for node in comm:
                node_communities[node] = i
        
        # Find nodes that influence multiple communities
        for node in G.nodes():
            if node not in node_communities:
                continue
                
            home_comm = node_communities[node]
            neighbors = list(G.neighbors(node))
            neighbor_comms = [node_communities.get(n, -1) for n in neighbors]
            unique_comms = set(c for c in neighbor_comms if c != -1 and c != home_comm)
            
            if len(unique_comms) >= 2:  # Node influences at least 2 communities besides its own
                cross_comm_influence = {
                    'node': node,
                    'home_community': home_comm,
                    'influenced_communities': list(unique_comms),
                    'betweenness': metrics[node].get('betweenness', 0),
                    'influencer_score': metrics[node].get('influencer_score', 0)
                }
                cross_community_influencers.append(cross_comm_influence)
    
    # Sort by number of influenced communities, then by betweenness
    cross_community_influencers.sort(
        key=lambda x: (len(x['influenced_communities']), x['betweenness']), 
        reverse=True
    )
    
    results['cross_community_influencers'] = cross_community_influencers[:10]  # Top 10
    
    return results


# ───────────────────── 2. Temporal Evolution Study ─────────────────────

def analyze_temporal_evolution(G_temporal, edges_df, window_count=TIME_WINDOWS):
    """
    Analyze the evolution of the network over time, identifying key changes
    in community structure and influence patterns.
    
    Parameters:
    - G_temporal: NetworkX graph with temporal information
    - edges_df: DataFrame containing edge information with timestamps
    - window_count: Number of time windows to analyze
    
    Returns:
    - Dictionary with temporal analysis results
    """
    results = {}
    
    # First, ensure we have temporal data
    if edges_df is None or 'timestamp' not in edges_df.columns:
        return {"error": "No temporal data available"}
    
    # Create time windows
    edges_df = edges_df.sort_values('timestamp')
    start_time = edges_df['timestamp'].min()
    end_time = edges_df['timestamp'].max()
    time_range = end_time - start_time
    window_size = time_range / window_count
    
    # Initialize window boundaries
    window_boundaries = [start_time + i * window_size for i in range(window_count + 1)]
    
    # Initialize data structures for tracking evolution
    window_graphs = []
    window_metrics = []
    window_communities = []
    window_influencers = []
    window_stats = []
    
    # For each time window, build a graph and analyze
    for i in range(window_count):
        window_start = window_boundaries[i]
        window_end = window_boundaries[i+1]
        
        # Filter edges for this time window
        window_edges = edges_df[(edges_df['timestamp'] >= window_start) & 
                               (edges_df['timestamp'] < window_end)]
        
        # Build graph for this window
        G_window = nx.Graph()
        for _, row in window_edges.iterrows():
            if row['source'] and row['target']:
                if not G_window.has_edge(row['source'], row['target']):
                    G_window.add_edge(row['source'], row['target'], weight=1)
                else:
                    G_window[row['source']][row['target']]['weight'] += 1
        
        # Skip if graph is too small for meaningful analysis
        if G_window.number_of_nodes() < 5 or G_window.number_of_edges() < 5:
            window_graphs.append(G_window)
            window_metrics.append({})
            window_communities.append({})
            window_influencers.append({})
            window_stats.append({
                'window_start': window_start,
                'window_end': window_end,
                'nodes': G_window.number_of_nodes(),
                'edges': G_window.number_of_edges(),
                'too_small': True
            })
            continue
        
        # Compute metrics and communities for this window
        try:
            node_metrics, _, network_metrics = novel_metrics(G_window)
            communities, community_eval, _ = community_scan(G_window)
            
            # Track top influencers in this window
            top_influencers = {}
            for measure in ['degree', 'betweenness', 'eigenvector']:
                if all(measure in node_metrics[node] for node in G_window.nodes()):
                    ranked_nodes = sorted(G_window.nodes(), 
                                         key=lambda n: node_metrics[n].get(measure, 0), 
                                         reverse=True)
                    top_influencers[measure] = [(node, node_metrics[node].get(measure, 0)) 
                                              for node in ranked_nodes[:5]]
            
            # Add to results
            window_graphs.append(G_window)
            window_metrics.append(node_metrics)
            window_communities.append(communities)
            window_influencers.append(top_influencers)
            
            # Compute aggregate statistics
            window_stats.append({
                'window_start': window_start,
                'window_end': window_end,
                'nodes': G_window.number_of_nodes(),
                'edges': G_window.number_of_edges(),
                'density': nx.density(G_window),
                'avg_clustering': network_metrics.get('avg_clustering', 0),
                'modularity': community_eval.get('louvain', {}).get('modularity', 0),
                'communities': community_eval.get('louvain', {}).get('communities', 0),
                'too_small': False
            })
        except Exception as e:
            window_graphs.append(G_window)
            window_metrics.append({})
            window_communities.append({})
            window_influencers.append({})
            window_stats.append({
                'window_start': window_start,
                'window_end': window_end,
                'nodes': G_window.number_of_nodes(),
                'edges': G_window.number_of_edges(),
                'error': str(e)
            })
    
    results['window_stats'] = window_stats
    
    # Analyze community evolution across windows
    community_evolution = []
    
    # Track nodes that moved between communities
    node_community_timeline = defaultdict(list)
    
    # For each window, track community assignments
    for window_idx, communities_dict in enumerate(window_communities):
        if 'louvain' not in communities_dict:
            continue
            
        for comm_idx, comm in enumerate(communities_dict['louvain']):
            for node in comm:
                node_community_timeline[node].append((window_idx, comm_idx))
    
    # Find nodes that changed communities
    community_migrants = []
    for node, timeline in node_community_timeline.items():
        if len(timeline) > 1:
            # Check if community changed
            community_changes = []
            prev_window, prev_comm = timeline[0]
            
            for window_idx, comm_idx in timeline[1:]:
                if comm_idx != prev_comm:
                    community_changes.append((prev_window, window_idx, prev_comm, comm_idx))
                    prev_comm = comm_idx
                prev_window = window_idx
                
            if community_changes:
                community_migrants.append({
                    'node': node,
                    'changes': community_changes
                })
    
    results['community_migrants'] = sorted(community_migrants, 
                                          key=lambda x: len(x['changes']), 
                                          reverse=True)[:20]  # Top 20 migrators
    
    # Analyze community stability
    community_stability = []
    
    for i in range(window_count-1):
        # Skip if either window is missing communities
        if ('louvain' not in window_communities[i] or 
            'louvain' not in window_communities[i+1]):
            continue
            
        # Get communities for consecutive windows
        communities_i = window_communities[i]['louvain']
        communities_i1 = window_communities[i+1]['louvain']
        
        # Create sets of nodes in each community
        comm_sets_i = [set(comm) for comm in communities_i]
        comm_sets_i1 = [set(comm) for comm in communities_i1]
        
        # Calculate Jaccard similarity between communities
        similarities = []
        
        for comm1 in comm_sets_i:
            if not comm1:
                continue
                
            best_match = 0
            for comm2 in comm_sets_i1:
                if not comm2:
                    continue
                    
                # Calculate Jaccard similarity
                intersection = len(comm1.intersection(comm2))
                union = len(comm1.union(comm2))
                
                if union > 0:
                    similarity = intersection / union
                    best_match = max(best_match, similarity)
            
            similarities.append(best_match)
        
        # Calculate average stability score
        if similarities:
            stability = sum(similarities) / len(similarities)
        else:
            stability = 0
            
        community_stability.append({
            'window_pair': (i, i+1),
            'stability_score': stability
        })
    
    results['community_stability'] = community_stability
    
    # Track influencer consistency
    influencer_consistency = {}
    
    # For each influence measure, track how consistently nodes appear in top ranks
    for measure in ['degree', 'betweenness', 'eigenvector']:
        # Count appearances in top 5
        appearance_count = Counter()
        
        for window_idx, influencers in enumerate(window_influencers):
            if measure not in influencers:
                continue
                
            for node, _ in influencers[measure]:
                appearance_count[node] += 1
        
        # Calculate consistency score (appearances / eligible windows)
        eligible_windows = sum(1 for inf in window_influencers if measure in inf)
        
        if eligible_windows > 0:
            consistency_scores = {node: count / eligible_windows 
                                 for node, count in appearance_count.items()}
            
            # Sort by consistency score
            sorted_consistency = sorted(consistency_scores.items(), 
                                      key=lambda x: x[1], 
                                      reverse=True)
            
            influencer_consistency[measure] = sorted_consistency[:10]  # Top 10
    
    results['influencer_consistency'] = influencer_consistency
    
    # Noteworthy events detection
    noteworthy_events = []
    
    # Look for significant changes between consecutive windows
    for i in range(1, len(window_stats)):
        current = window_stats[i]
        previous = window_stats[i-1]
        
        # Skip windows with errors or that are too small
        if (current.get('too_small', False) or 
            previous.get('too_small', False) or
            'error' in current or 'error' in previous):
            continue
        
        events = []
        
        # Check for significant growth/decline in nodes
        if current['nodes'] > 0 and previous['nodes'] > 0:
            node_change = (current['nodes'] - previous['nodes']) / previous['nodes']
            if abs(node_change) > 0.25:  # 25% change
                events.append({
                    'type': 'node_change',
                    'description': f"{'Increase' if node_change > 0 else 'Decrease'} in active contributors by {abs(node_change)*100:.1f}%",
                    'magnitude': abs(node_change)
                })
        
        # Check for significant changes in community structure
        if ('modularity' in current and 'modularity' in previous and
            current['modularity'] > 0 and previous['modularity'] > 0):
            mod_change = current['modularity'] - previous['modularity']
            if abs(mod_change) > 0.1:  # Significant modularity change
                events.append({
                    'type': 'modularity_change',
                    'description': f"{'Strengthening' if mod_change > 0 else 'Weakening'} of community structure",
                    'magnitude': abs(mod_change)
                })
        
        # Check for changes in community count
        if ('communities' in current and 'communities' in previous and
            current['communities'] > 0 and previous['communities'] > 0):
            comm_change = current['communities'] - previous['communities']
            if abs(comm_change) > 1:  # More than 1 community change
                events.append({
                    'type': 'community_count_change',
                    'description': f"{'Formation' if comm_change > 0 else 'Consolidation'} of communities: {abs(comm_change)} communities {'added' if comm_change > 0 else 'merged'}",
                    'magnitude': abs(comm_change)
                })
        
        if events:
            noteworthy_events.append({
                'window_pair': (i-1, i),
                'window_start': previous['window_end'],
                'window_end': current['window_end'],
                'events': events
            })
    
    results['noteworthy_events'] = noteworthy_events
    
    return results


# ───────────────────── 3. Practical Recommendations & Governance Toolkit ─────────────────────

def generate_governance_toolkit(network_results, community_results, temporal_results):
    """
    Generate practical recommendations and a governance toolkit based on
    network analysis findings.
    
    Parameters:
    - network_results: Results from network analysis
    - community_results: Results from community detection and influence analysis
    - temporal_results: Results from temporal evolution analysis
    
    Returns:
    - Dictionary with governance toolkit and recommendations
    """
    toolkit = {}
    
    # Extract metrics from results
    network_metrics = network_results.get('network_metrics', {})
    node_metrics = network_results.get('metrics', {})
    communities = network_results.get('communities', {})
    community_profiles = community_results.get('community_profiles', {})
    cross_comm_influencers = community_results.get('cross_community_influencers', [])
    
    # --- Growth and Sustainability Strategies ---
    growth_strategies = []
    
    # Check for power law degree distribution
    if network_metrics.get('power_law_alpha'):
        alpha = network_metrics['power_law_alpha']
        if 2 < alpha < 3:  # Scale-free network
            growth_strategies.append({
                'title': "Leverage the scale-free structure",
                'description': "Your network shows classic scale-free properties with preferential attachment. "
                              "New contributors are more likely to collaborate with visible, established contributors.",
                'actions': [
                    "Create 'Getting Started' guides that connect newcomers with mid-level contributors rather than overwhelming central hubs",
                    "Implement a mentorship program matching newcomers with mid-tier contributors",
                    "Make contribution pathways visible through documentation that shows collaboration networks"
                ],
                'rationale': f"Scale-free networks (power law α={alpha:.2f}) naturally grow through preferential attachment, "
                             "but can become vulnerable if hub nodes leave"
            })
    
    # Check for presence of structural roles
    role_counts = Counter()
    for node, metrics in node_metrics.items():
        if 'structural_role' in metrics:
            role_counts[metrics['structural_role']] += 1
    
    # If periphery is large relative to core, recommend onboarding strategies
    if role_counts.get('Peripheral', 0) > 2 * role_counts.get('Core', 0):
        growth_strategies.append({
            'title': "Convert peripheral contributors to regular contributors",
            'description': "Your network has a large periphery of occasional contributors that could become more engaged.",
            'actions': [
                "Create 'Second Contribution' incentives like badges or recognition",
                "Actively reach out to one-time contributors with specific follow-up tasks",
                "Track and reduce the 'time to first feedback' for newcomer contributions",
                "Set up automated first-issue suggestions for new contributors based on their initial work"
            ],
            'rationale': f"Network has {role_counts.get('Peripheral', 0)} peripheral nodes vs. "
                         f"{role_counts.get('Core', 0)} core nodes; research shows contributors who make a second "
                         f"contribution within 48 hours are much more likely to become regulars"
        })
    
    # Community-specific growth strategies
    if community_profiles:
        # Find small communities that could benefit from growth
        small_communities = [idx for idx, profile in community_profiles.items() 
                           if profile['size'] < 5 and profile['cohesion'] > 0.3]
        
        if small_communities:
            comms_str = ", ".join([str(idx) for idx in small_communities[:3]])
            if len(small_communities) > 3:
                comms_str += f" and {len(small_communities)-3} others"
                
            growth_strategies.append({
                'title': "Nurture promising small communities",
                'description': f"Communities {comms_str} show high cohesion but small size. "
                               "These could become important project areas with targeted growth.",
                'actions': [
                    "Create focused documentation for these specialized areas",
                    "Highlight these communities in project newsletters or meetings",
                    "Consider dedicated meetings or channels for these topic areas",
                    "Recruit experienced contributors with relevant expertise to join these communities"
                ],
                'rationale': "Small but cohesive communities often represent specialized expertise "
                             "that can drive innovation if properly supported"
            })
    
    toolkit['growth_strategies'] = growth_strategies
    
    # --- Conflict Management Framework ---
    conflict_strategies = []
    
    # Check for community polarization
    if 'inter_community_connections' in community_results:
        connections = community_results['inter_community_connections']
        if connections:
            # Calculate average connections between communities
            avg_connections = sum(connections.values()) / len(connections)
            
            # Find poorly connected community pairs
            low_connection_pairs = [(c1, c2) for (c1, c2), count in connections.items() 
                                  if count < 0.5 * avg_connections]
            
            if low_connection_pairs:
                conflict_strategies.append({
                    'title': "Bridge divided communities",
                    'description': "Some communities in your project have few connections, "
                                   "which may indicate siloed work or potential conflicts.",
                    'actions': [
                        "Create cross-community task forces for features that span multiple areas",
                        "Implement a review rotation system where members review PRs outside their community",
                        "Host technical exchange sessions between specific communities",
                        "Identify and support contributors who can serve as bridges"
                    ],
                    'rationale': "Network analysis shows weak connections between some communities, "
                                 "which may lead to divergent practices or communication barriers"
                })
    
    # Check for rich club phenomenon (potential power concentration)
    if network_results.get('richclub'):
        richclub = network_results['richclub']
        
        # Check if higher degree nodes show rich club effect
        high_k_values = [v for k, v in richclub.items() if k > 5]
        if high_k_values and any(v > 1.5 for v in high_k_values):
            conflict_strategies.append({
                'title': "Manage power concentration",
                'description': "Your project shows a 'rich club' effect where high-influence contributors "
                               "primarily collaborate with each other, which can lead to perceived exclusivity.",
                'actions': [
                    "Implement a 'sponsor' system where core members actively collaborate with newer contributors",
                    "Create transparent decision-making processes with published rationales",
                    "Rotate leadership roles for specific project areas",
                    "Institute regular retrospectives to surface concerns about project governance"
                ],
                'rationale': "Rich club analysis shows high-degree nodes collaborate preferentially with "
                             "each other, which can create an 'inner circle' perception"
            })
    
    # Analyze temporal stability for potential conflict patterns
    if 'community_stability' in temporal_results:
        stability = temporal_results['community_stability']
        if stability:
            avg_stability = sum(s['stability_score'] for s in stability) / len(stability)
            
            if avg_stability < 0.6:  # Low stability might indicate churn due to conflicts
                conflict_strategies.append({
                    'title': "Address community instability",
                    'description': "Your project shows significant community restructuring over time, "
                                   "which may indicate unresolved tensions or direction changes.",
                    'actions': [
                        "Institute clear code ownership policies to reduce territorial conflicts",
                        "Document architecture decisions and their rationales",
                        "Create an explicit conflict resolution process with neutral mediators",
                        "Regular health surveys to catch issues before they cause contributor departures"
                    ],
                    'rationale': f"Temporal analysis shows community stability score of {avg_stability:.2f}, "
                                 f"indicating significant reorganization of collaboration patterns"
                })
    
    toolkit['conflict_management'] = conflict_strategies
    
    # --- Knowledge Sharing Framework ---
    knowledge_strategies = []
    
    # Check for bus factor issues (over-reliance on few contributors)
    if 'top_contributors' in community_results:
        top_degree = community_results['top_contributors'].get('degree', [])
        if top_degree:
            # Calculate what percentage of connections involve the top 3 contributors
            total_edges = network_results['graph'].number_of_edges()
            top3_edges = sum(deg for _, deg in top_degree[:3])
            
            if top3_edges / (2 * total_edges) > 0.3:  # Top 3 involved in >30% of collaborations
                knowledge_strategies.append({
                    'title': "Reduce bus factor risk",
                    'description': "Your project shows high dependence on a small number of key contributors, "
                                   "creating risk if these individuals become unavailable.",
                    'actions': [
                        "Create CODEOWNERS files that require multiple reviewers",
                        "Implement pair programming or shadowing for critical subsystems",
                        "Document architectural decisions and implicit knowledge",
                        "Create onboarding paths to critical subsystems with graduated responsibility"
                    ],
                    'rationale': f"Network centrality shows top 3 contributors are involved in "
                                 f"{(top3_edges/(2*total_edges)*100):.1f}% of all collaborations"
                })
    
    # Bridge node knowledge transfer
    if cross_comm_influencers:
        knowledge_strategies.append({
            'title': "Leverage cross-community knowledge brokers",
            'description': "Your project has valuable contributors who bridge multiple communities "
                           "and can facilitate knowledge transfer.",
            'actions': [
                f"Ask {cross_comm_influencers[0]['node']} to help document interfaces between communities",
                "Create a 'technical ambassadors' program with bridge contributors",
                "Host knowledge-sharing sessions led by cross-community contributors",
                "Set up explicit mentoring relationships between bridge nodes and newer contributors"
            ],
            'rationale': "Network analysis identified valuable bridge nodes who have connections "
                         "across multiple communities and can transfer best practices"
        })
    
    # Analyze community expertise patterns
    if community_profiles:
        # Find densely connected communities (specialized knowledge areas)
        specialized_communities = [idx for idx, profile in community_profiles.items() 
                                 if profile['density'] > 0.5 and profile['size'] >= 5]
        
        if specialized_communities:
            knowledge_strategies.append({
                'title': "Document specialized knowledge areas",
                'description': "Your project has several densely-connected subcommunities that likely "
                               "represent specialized expertise areas.",
                'actions': [
                    "Create expertise directories mapping topics to contributors",
                    "Develop specialized documentation for each knowledge area",
                    "Institute a 'community spotlight' in project communication",
                    "Create onboarding paths specific to each specialized area"
                ],
                'rationale': f"Network analysis identified {len(specialized_communities)} communities "
                             f"with high density, indicating specialized knowledge clusters"
            })
    
    toolkit['knowledge_sharing'] = knowledge_strategies
    
    # --- Contributor Retention Strategies ---
    retention_strategies = []
    
    # Check for declining participation trends
    if 'window_stats' in temporal_results:
        stats = temporal_results['window_stats']
        
        # Filter out windows that are too small
        valid_stats = [s for s in stats if not s.get('too_small', False) and 'error' not in s]
        
        if len(valid_stats) >= 3:  # Need at least 3 valid windows for trend
            early_nodes = sum(s['nodes'] for s in valid_stats[:len(valid_stats)//3]) / (len(valid_stats)//3)
            late_nodes = sum(s['nodes'] for s in valid_stats[-len(valid_stats)//3:]) / (len(valid_stats)//3)
            
            if late_nodes < 0.8 * early_nodes:  # 20% decline
                retention_strategies.append({
                    'title': "Address declining participation",
                    'description': "Your project shows a declining trend in active contributors, "
                                   "which may indicate retention issues.",
                    'actions': [
                        "Survey recently inactive contributors to understand departure reasons",
                        "Create recognition programs to acknowledge consistent contributors",
                        "Streamline contribution processes to reduce friction",
                        "Establish clear growth paths for contributors seeking to increase involvement"
                    ],
                    'rationale': f"Temporal analysis shows {((early_nodes-late_nodes)/early_nodes*100):.1f}% "
                                 f"decline in active contributors over the analyzed period"
                })
    
    # Check for inconsistent contributor participation
    if 'influencer_consistency' in temporal_results:
        consistency = temporal_results['influencer_consistency']
        if 'degree' in consistency:
            # Calculate average consistency score
            avg_consistency = sum(score for _, score in consistency['degree'][:10]) / min(10, len(consistency['degree']))
            
            if avg_consistency < 0.7:  # Low consistency
                retention_strategies.append({
                    'title': "Improve contributor consistency",
                    'description': "Your project shows irregular participation patterns even among active contributors, "
                                   "which may indicate barriers to sustained engagement.",
                    'actions': [
                        "Create a predictable release and maintenance schedule",
                        "Implement regular check-in mechanisms for contributors",
                        "Identify and remove friction points in the contribution process",
                        "Consider time zone and regional factors in project scheduling"
                    ],
                    'rationale': f"Temporal analysis shows key contributors have an average consistency "
                                 f"score of {avg_consistency:.2f}, indicating irregular participation"
                })
    
    # Look for community stability issues related to retention
    if 'noteworthy_events' in temporal_results:
        events = temporal_results['noteworthy_events']
        node_decline_events = []
        
        for event_group in events:
            for event in event_group['events']:
                if event['type'] == 'node_change' and event['magnitude'] < 0:  # Negative change = decline
                    node_decline_events.append(event)
        
        if len(node_decline_events) > 1:  # Multiple decline events
            retention_strategies.append({
                'title': "Investigate contribution drop-off points",
                'description': "Your project shows specific periods of contributor decline that may "
                               "correspond to project events or external factors.",
                'actions': [
                    "Review major changes preceding contribution drops",
                    "Analyze issue/PR backlogs during decline periods",
                    "Create community health metrics to monitor engagement",
                    "Review project governance changes during decline periods"
                ],
                'rationale': f"Temporal analysis identified {len(node_decline_events)} periods of "
                             f"significant contributor decline that warrant investigation"
            })
    
    # Check for long-term community health
    if community_results.get('community_profiles') and temporal_results.get('window_stats'):
        # Check if any communities have disappeared over time
        if len(temporal_results['window_stats']) > 3:
            early_window = next((s for s in temporal_results['window_stats'] 
                              if not s.get('too_small', False) and 'communities' in s), None)
            late_window = next((s for s in reversed(temporal_results['window_stats']) 
                             if not s.get('too_small', False) and 'communities' in s), None)
            
            if early_window and late_window and early_window['communities'] > late_window['communities']:
                retention_strategies.append({
                    'title': "Revitalize lost community areas",
                    'description': "Your project has lost specialized contributor communities over time, "
                                   "which may indicate neglected project areas.",
                    'actions': [
                        "Identify and re-prioritize abandoned project areas",
                        "Create 'community reboot' initiatives for dormant areas",
                        "Review successful communities for practices to apply to declining ones",
                        "Consider formal sunsetting process for truly obsolete areas"
                    ],
                    'rationale': f"Community analysis shows a reduction from {early_window['communities']} to "
                                 f"{late_window['communities']} distinct contributor communities"
                })
    
    toolkit['retention_strategies'] = retention_strategies
    
    # --- Decision-Making Framework ---
    decision_strategies = []
    
    # Analyze network structure for decision bottlenecks
    if node_metrics:
        betweenness_scores = [(node, metrics.get('betweenness', 0)) 
                            for node, metrics in node_metrics.items()]
        betweenness_scores.sort(key=lambda x: x[1], reverse=True)
        
        # Check if top betweenness nodes are critical bottlenecks
        if betweenness_scores and betweenness_scores[0][1] > 0:
            top_score = betweenness_scores[0][1]
            median_score = betweenness_scores[len(betweenness_scores)//2][1] if len(betweenness_scores) > 1 else 0
            
            if median_score > 0 and top_score / median_score > 10:  # High inequality in betweenness
                decision_strategies.append({
                    'title': "Address decision bottlenecks",
                    'description': f"Your project shows high centralization around {betweenness_scores[0][0]}, "
                                   f"who may be a bottleneck for project decisions.",
                    'actions': [
                        "Implement a time-boxed decision process with explicit delegation",
                        "Create domain-specific authority roles with clear decision scope",
                        "Train additional reviewers to reduce approval bottlenecks",
                        "Document decision-making processes with escalation paths"
                    ],
                    'rationale': f"Network analysis shows top contributor has {top_score/median_score:.1f}x "
                                 f"the median betweenness centrality, indicating potential bottlenecks"
                })
    
    # Community-based decision strategies
    if community_profiles:
        # Check for isolated decision-making in communities
        isolated_communities = [idx for idx, profile in community_profiles.items() 
                             if len(profile.get('bridges', [])) < 2 and profile['size'] > 5]
        
        if isolated_communities:
            decision_strategies.append({
                'title': "Improve cross-community decision coordination",
                'description': "Your project has communities with limited external connections, "
                               "which can lead to misaligned decisions and duplicated work.",
                'actions': [
                    "Create a cross-community architecture council",
                    "Implement project-wide RFC processes for significant changes",
                    "Establish regular cross-community sync meetings",
                    "Create shared roadmaps with explicit cross-community dependencies"
                ],
                'rationale': f"Network analysis identified {len(isolated_communities)} communities "
                             f"with limited bridges to other groups"
            })
    
    # Check for project governance clarity based on community structure
    if 'algorithm_comparison' in community_results:
        alg_comparison = community_results['algorithm_comparison']
        
        # If community detection algorithms disagree significantly, governance may be unclear
        if len(alg_comparison) >= 2:
            counts = [details['count'] for alg, details in alg_comparison.items()]
            max_count = max(counts)
            min_count = min(counts)
            
            if max_count > 2 * min_count:  # Big difference in community counts
                decision_strategies.append({
                    'title': "Clarify project governance structure",
                    'description': "Your project shows ambiguous community structure, which may "
                                   "indicate unclear governance and decision processes.",
                    'actions': ["Document and publish project governance model",
                        "Create explicit decision-making frameworks for different types of changes",
                        "Establish clear escalation paths for contentious decisions",
                        "Create CONTRIBUTING.md with explicit governance information"
                    ],
                    'rationale': f"Community detection algorithms show inconsistent results "
                                 f"({min_count}-{max_count} communities), suggesting unclear structure"
                })
    
    toolkit['decision_making'] = decision_strategies
    
    # --- Create Final Report ---
    
    # Generate executive summary
    top_strategies = []
    for category in ['growth_strategies', 'conflict_management', 'knowledge_sharing', 
                     'retention_strategies', 'decision_making']:
        if category in toolkit and toolkit[category]:
            # Take the first strategy from each category
            top_strategies.append({
                'category': category.replace('_', ' ').title(),
                'strategy': toolkit[category][0]['title'],
                'description': toolkit[category][0]['description']
            })
    
    # Generate markdown report
    report = "# OSS Project Governance Toolkit\n\n"
    
    report += "## Executive Summary\n\n"
    report += "Based on comprehensive network analysis of your project's collaboration patterns, "
    report += "these are the top recommended strategies for enhancing project governance:\n\n"
    
    for i, strategy in enumerate(top_strategies):
        report += f"{i+1}. **{strategy['category']}**: {strategy['strategy']} - {strategy['description']}\n\n"
    
    # Add detailed sections
    for category, strategies in toolkit.items():
        if not strategies:
            continue
            
        report += f"## {category.replace('_', ' ').title()}\n\n"
        
        for i, strategy in enumerate(strategies):
            report += f"### {i+1}. {strategy['title']}\n\n"
            report += f"{strategy['description']}\n\n"
            report += "**Recommended Actions:**\n\n"
            
            for action in strategy['actions']:
                report += f"- {action}\n"
            
            report += f"\n**Rationale:** {strategy['rationale']}\n\n"
    
    toolkit['report'] = report
    
    return toolkit


# ───────────────────── 4. Enhanced Visualization Functions ─────────────────────

def visualize_community_evolution(temporal_results):
    """
    Create advanced visualizations showing the evolution of communities over time.
    
    Parameters:
    - temporal_results: Results from temporal evolution analysis
    
    Returns:
    - Dictionary with visualization figures
    """
    figs = {}
    
    # Extract time window data
    if 'window_stats' not in temporal_results:
        return {'error': 'No temporal window stats available'}
        
    window_stats = temporal_results['window_stats']
    
    # Filter out invalid windows
    valid_windows = [w for w in window_stats if not w.get('too_small', False) and 'error' not in w]
    
    if len(valid_windows) < 2:
        return {'error': 'Not enough valid time windows for visualization'}
    
    # --- 1. Community Size Evolution Plot ---
    plt.figure(figsize=(12, 7))
    
    # Extract time labels and community counts
    time_labels = [w.get('label', f"Window {i+1}") for i, w in enumerate(valid_windows)]
    community_counts = [w.get('communities', 0) for w in valid_windows]
    node_counts = [w.get('nodes', 0) for w in valid_windows]
    edge_counts = [w.get('edges', 0) for w in valid_windows]
    
    # Plot community count line
    plt.plot(time_labels, community_counts, 'o-', linewidth=2, color='#1f77b4', label='Communities')
    
    # Create second y-axis for nodes/edges
    ax2 = plt.gca().twinx()
    ax2.plot(time_labels, node_counts, 's--', color='#ff7f0e', label='Active Contributors')
    ax2.plot(time_labels, edge_counts, '^:', color='#2ca02c', label='Collaborations')
    
    # Add labels and legend
    plt.gca().set_xlabel('Time Period', fontsize=12)
    plt.gca().set_ylabel('Number of Communities', fontsize=12)
    ax2.set_ylabel('Contributors & Collaborations', fontsize=12)
    
    # Combine legends from both axes
    lines1, labels1 = plt.gca().get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.1),
               fancybox=True, shadow=True, ncol=3)
    
    plt.title('Project Evolution Over Time', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # Save figure
    figs['community_evolution'] = plt.gcf()
    
    # --- 2. Community Stability Alluvial/Sankey Diagram ---
    if 'community_tracking' in temporal_results:
        tracking = temporal_results['community_tracking']
        
        if len(tracking) >= 2:
            plt.figure(figsize=(14, 8))
            
            # Create a Sankey diagram of community evolution
            try:
                # Extract nodes and links
                labels = []
                source = []
                target = []
                value = []
                node_colors = []
                
                # Generate distinct community colors
                cmap = plt.cm.viridis
                
                # Process each time transition
                for t, transition in enumerate(tracking):
                    # Add source nodes (previous window communities)
                    if 'transitions' in transition:
                        for prev_id, destinations in transition['transitions'].items():
                            # Source node label
                            src_label = f"T{t}-C{prev_id}"
                            if src_label not in labels:
                                labels.append(src_label)
                                node_colors.append(cmap(prev_id / max(1, max(transition['transitions'].keys()))))
                            
                            src_idx = labels.index(src_label)
                            
                            # Process destination communities
                            for dest_id, flow in destinations.items():
                                # Target node label
                                tgt_label = f"T{t+1}-C{dest_id}"
                                if tgt_label not in labels:
                                    labels.append(tgt_label)
                                    node_colors.append(cmap(dest_id / max(1, max(destinations.keys()))))
                                
                                tgt_idx = labels.index(tgt_label)
                                
                                # Add link
                                source.append(src_idx)
                                target.append(tgt_idx)
                                value.append(flow)
                
                # Create Sankey diagram
                if source and target and value:
                    fig = go.Figure(data=[go.Sankey(
                        node=dict(
                            pad=15,
                            thickness=20,
                            line=dict(color="black", width=0.5),
                            label=labels,
                            color=[f"rgba({int(c[0]*255)},{int(c[1]*255)},{int(c[2]*255)},0.8)" for c in node_colors]
                        ),
                        link=dict(
                            source=source,
                            target=target,
                            value=value
                        )
                    )])
                    
                    fig.update_layout(
                        title_text="Community Evolution Flow",
                        font_size=12
                    )
                    
                    figs['community_flow'] = fig
            except Exception as e:
                print(f"Error creating Sankey diagram: {e}")
                pass
    
    # --- 3. Influencer Consistency Visualization ---
    if 'influencer_consistency' in temporal_results:
        consistency = temporal_results['influencer_consistency']
        
        if 'degree' in consistency and consistency['degree']:
            plt.figure(figsize=(10, 8))
            
            # Get top 10 influencers and their consistency scores
            top_influencers = consistency['degree'][:10]
            
            names = [inf['node'] for inf in top_influencers]
            scores = [inf['consistency'] for inf in top_influencers]
            
            # Create horizontal bar chart
            y_pos = np.arange(len(names))
            bars = plt.barh(y_pos, scores, height=0.7)
            
            # Color bars by consistency score
            cmap = plt.cm.RdYlGn
            for i, bar in enumerate(bars):
                bar.set_color(cmap(scores[i]))
            
            # Add labels
            plt.yticks(y_pos, names)
            plt.xlabel('Consistency Score (higher = more consistent)')
            plt.title('Top Influencer Consistency Over Time')
            
            # Add colorbar
            sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(0, 1))
            sm.set_array([])
            cbar = plt.colorbar(sm)
            cbar.set_label('Consistency Level', rotation=270, labelpad=20)
            
            plt.tight_layout()
            figs['influencer_consistency'] = plt.gcf()
    
    # --- 4. Network Health Dashboard ---
    if 'health_metrics' in temporal_results:
        health = temporal_results['health_metrics']
        
        if health and len(health) > 1:
            # Create figure with subplots
            fig, axs = plt.subplots(2, 2, figsize=(14, 10))
            
            # Extract time points and metrics
            times = [h.get('window_label', f"T{i}") for i, h in enumerate(health)]
            
            # Metrics to plot
            metrics = {
                'modularity': [h.get('modularity', None) for h in health],
                'avg_clustering': [h.get('clustering', None) for h in health],
                'density': [h.get('density', None) for h in health],
                'contributor_ratio': [h.get('new_contributor_ratio', None) for h in health]
            }
            
            # Plot each metric
            for i, (metric, values) in enumerate(metrics.items()):
                row, col = i // 2, i % 2
                ax = axs[row, col]
                
                # Filter out None values
                valid_times = []
                valid_values = []
                for t, v in zip(times, values):
                    if v is not None:
                        valid_times.append(t)
                        valid_values.append(v)
                
                if valid_values:
                    ax.plot(valid_times, valid_values, 'o-', linewidth=2)
                    
                    # Add trend line
                    if len(valid_values) > 2:
                        z = np.polyfit(range(len(valid_values)), valid_values, 1)
                        p = np.poly1d(z)
                        ax.plot(valid_times, p(range(len(valid_values))), 'r--', alpha=0.6)
                    
                    # Format plot
                    ax.set_title(metric.replace('_', ' ').title())
                    ax.grid(True, alpha=0.3)
                    ax.set_xticks(valid_times)
                    ax.set_xticklabels(valid_times, rotation=45)
            
            plt.tight_layout()
            figs['network_health'] = plt.gcf()
    
    return figs


def create_community_profile_visualization(community_results):
    """
    Create visualizations showing the profiles of different communities.
    
    Parameters:
    - community_results: Results from community detection and influence analysis
    
    Returns:
    - Dictionary with visualization figures
    """
    figs = {}
    
    # Check if we have community profiles
    if 'community_profiles' not in community_results:
        return {'error': 'No community profiles available'}
    
    profiles = community_results['community_profiles']
    
    if not profiles:
        return {'error': 'Empty community profiles'}
    
    # --- 1. Community Comparison Radar Chart ---
    # Select metrics to display
    metrics = ['size', 'density', 'cohesion', 'isolation', 'diversity']
    
    # Get communities with all metrics
    valid_communities = {}
    for comm_id, profile in profiles.items():
        if all(m in profile for m in metrics):
            valid_communities[comm_id] = profile
    
    if valid_communities:
        # Select top communities by size (up to 5)
        top_communities = sorted(valid_communities.items(), 
                                key=lambda x: x[1]['size'], 
                                reverse=True)[:5]
        
        # Create figure with multiple traces
        fig = go.Figure()
        
        for comm_id, profile in top_communities:
            # Extract values for this community
            values = [profile[m] for m in metrics]
            # Close the polygon
            values.append(values[0])
            metrics_plot = metrics + [metrics[0]]
            
            # Add trace
            fig.add_trace(go.Scatterpolar(
                r=values,
                theta=metrics_plot,
                fill='toself',
                name=f'Community {comm_id}'
            ))
        
        fig.update_layout(
            polar=dict(
                radialaxis=dict(
                    visible=True,
                    range=[0, 1]
                )
            ),
            showlegend=True,
            title='Community Profile Comparison'
        )
        
        figs['community_radar'] = fig
    
    # --- 2. Community Network Visualization ---
    if 'community_networks' in community_results:
        comm_networks = community_results['community_networks']
        
        # Create a figure with subgraphs
        if comm_networks:
            fig, axs = plt.subplots(1, min(4, len(comm_networks)), 
                                  figsize=(16, 4), 
                                  squeeze=False)
            
            # Get top communities by size
            top_comms = sorted(comm_networks.items(), 
                             key=lambda x: len(x[1].nodes()), 
                             reverse=True)[:4]
            
            for i, (comm_id, G_comm) in enumerate(top_comms):
                if i >= 4:  # Only show up to 4 communities
                    break
                    
                ax = axs[0, i]
                
                # Draw the network
                pos = nx.spring_layout(G_comm, seed=42)
                nx.draw_networkx_nodes(G_comm, pos, node_size=50, node_color=PALETTE[i], alpha=0.8, ax=ax)
                nx.draw_networkx_edges(G_comm, pos, width=0.5, alpha=0.5, ax=ax)
                
                # Only label nodes with high degree
                top_nodes = sorted(G_comm.nodes(), key=lambda n: G_comm.degree(n), reverse=True)[:3]
                labels = {n: n for n in top_nodes}
                nx.draw_networkx_labels(G_comm, pos, labels=labels, font_size=8, ax=ax)
                
                ax.set_title(f"Community {comm_id}")
                ax.axis('off')
            
            plt.tight_layout()
            figs['community_networks'] = plt.gcf()
    
    # --- 3. Cross-Community Connections Heatmap ---
    if 'inter_community_connections' in community_results:
        connections = community_results['inter_community_connections']
        
        if connections:
            # Create a square matrix of community connections
            comm_ids = sorted(set([c1 for (c1, c2) in connections.keys()] + 
                               [c2 for (c1, c2) in connections.keys()]))
            
            # Initialize matrix
            n = len(comm_ids)
            matrix = np.zeros((n, n))
            
            # Fill matrix
            for (c1, c2), count in connections.items():
                i = comm_ids.index(c1)
                j = comm_ids.index(c2)
                matrix[i, j] = count
                matrix[j, i] = count  # Symmetric
            
            # Create heatmap
            plt.figure(figsize=(10, 8))
            sns.heatmap(matrix, annot=True, fmt=".0f", cmap="YlGnBu",
                      xticklabels=[f"C{c}" for c in comm_ids],
                      yticklabels=[f"C{c}" for c in comm_ids])
            
            plt.title("Cross-Community Connections")
            plt.tight_layout()
            figs['community_connections'] = plt.gcf()
    
    return figs


# ───────────────────── 5. Integration: Main Function ─────────────────────


def run_comprehensive_analysis(repos, save_path=None, verbose=True):
    # ────────────────── existing code (unchanged) ──────────────────
    base_results   = analyze_github_network(repos, save_path, verbose)
    
    community_res  = analyze_communities_and_influence(
                        base_results['graph'],
                        base_results['metrics'],
                        base_results['communities'])
    temporal_res   = analyze_temporal_evolution(
                        base_results['temporal_graph'],
                        base_results['edges_df'])
    governance_res = generate_governance_toolkit(
                        base_results, community_res, temporal_res)
    comm_viz       = create_community_profile_visualization(community_res)
    temp_viz       = visualize_community_evolution(temporal_res)

    # ───────────── NEW – run the *reporting / extra-viz* helpers ─────────────
    # pass only the directory, they will write "influencer_report.md" under it
    influencer_report_md = generate_influencer_report(
        community_res,
        save_path=save_path)

    community_health_md  = generate_community_health_report(
        community_res,
        temporal_res,
        save_path=save_path)

    # collect everything we need for governance recommendations
    rec_input = {
        **base_results,
        'community_results': community_res,
        'temporal_results': temporal_res,
        'network_metrics':  base_results.get('network_metrics', {})
    }
    governance_recs_md = generate_governance_recommendations(
        rec_input,
        save_path=save_path)

    # Influencer network figure
    influencer_fig = visualize_influencer_network(
        base_results['graph'],
        base_results['metrics'],
        base_results.get('node_to_comm', None),
        top_n=20)
    if save_path:
        influencer_fig.savefig(f"{save_path}/influencer_network.png",
                               dpi=300, bbox_inches="tight")

    # ───────────────── assemble everything to return ─────────────────
    comprehensive_results = {
        **base_results,
        'community_analysis'      : community_res,
        'temporal_analysis'       : temporal_res,
        'governance_toolkit'      : governance_res,
        'community_visualizations': comm_viz,
        'temporal_visualizations' : temp_viz,
        'influencer_report'       : influencer_report_md,
        'community_health_report' : community_health_md,
        'governance_recommendations_report': governance_recs_md,
        'influencer_network_fig'  : influencer_fig
    }
    
    # Save results if requested
    if save_path:
        try:
            # Save visualizations
            for viz_name, fig in comprehensive_results.get('community_visualizations', {}).items():
                if hasattr(fig, 'savefig'):
                    fig.savefig(f"{save_path}/community_{viz_name}.png", dpi=300, bbox_inches='tight')
            for viz_name, fig in comprehensive_results.get('temporal_visualizations', {}).items():
                if hasattr(fig, 'savefig'):
                    fig.savefig(f"{save_path}/temporal_{viz_name}.png", dpi=300, bbox_inches='tight')
            
            # Save governance toolkit report 
            # (note: governance_res is your toolkit dict)
            if 'report' in governance_res:
                with open(f"{save_path}/governance_toolkit.md", 'w') as f:
                    f.write(governance_res['report'])
            
            # Save comprehensive_results.json
            import json
            serializable = {}
            import datetime
            for k, v in comprehensive_results.items():
                if k in ['graph','temporal_graph','community_visualizations','temporal_visualizations']:
                    continue

                # Already JSON‐friendly types
                if isinstance(v, (dict, list, str, int, float, bool, type(None))):
                    serializable[k] = v

                # Convert Timestamps and datetimes to ISO format
                
                elif isinstance(v, (pd.Timestamp, datetime.datetime, datetime.date)):
                    serializable[k] = v.isoformat()

                # Fallback: stringify everything else
                else:
                    serializable[k] = str(v)

            with open(f"{save_path}/comprehensive_results.json", 'w') as f:
                json.dump(serializable, f, indent=2, default=str)
        except Exception as e:
            if verbose:
                print(f"Error saving results: {e}")
    
    return comprehensive_results


# ───────────────────── 6. Additional Reporting Functions ─────────────────────

def generate_influencer_report(community_results, save_path=None):
    """
    Generate a specialized report focusing on key influencers and their roles.
    
    Parameters:
    - community_results: Results from community detection and influence analysis
    - save_path: Path to save the report
    
    Returns:
    - Markdown report as a string
    """
    if 'top_contributors' not in community_results:
        return "No contributor data available for influencer report."
    
    top = community_results['top_contributors']
    
    # Create the report
    report = "# Key Influencers Analysis\n\n"
    
    report += "## Project Influence Landscape\n\n"
    
    # Add top contributors by different metrics
    if 'degree' in top:
        report += "### Most Connected Contributors\n\n"
        report += "These contributors collaborate with the most people:\n\n"
        for i, (node, value) in enumerate(top['degree'][:10]):
            report += f"{i+1}. **{node}** - Connected to {value} contributors\n"
        report += "\n"
    
    if 'betweenness' in top:
        report += "### Bridge Contributors\n\n"
        report += "These contributors connect different parts of the project:\n\n"
        for i, (node, value) in enumerate(top['betweenness'][:10]):
            report += f"{i+1}. **{node}** - Betweenness centrality: {value:.2f}\n"
        report += "\n"
    
    if 'eigenvector' in top:
        report += "### Core Contributors\n\n"
        report += "These contributors are connected to other well-connected contributors:\n\n"
        for i, (node, value) in enumerate(top['eigenvector'][:10]):
            report += f"{i+1}. **{node}** - Eigenvector centrality: {value:.2f}\n"
        report += "\n"
    
    if 'closeness' in top:
        report += "### Central Contributors\n\n"
        report += "These contributors have short paths to many others:\n\n"
        for i, (node, value) in enumerate(top['closeness'][:10]):
            report += f"{i+1}. **{node}** - Closeness centrality: {value:.2f}\n"
        report += "\n"
    
    # Add cross-community influencers
    if 'cross_community_influencers' in community_results:
        report += "## Cross-Community Influencers\n\n"
        report += "These contributors bridge multiple communities and transfer knowledge:\n\n"
        
        for i, info in enumerate(community_results['cross_community_influencers'][:10]):
            communities = info.get('communities', [])
            communities_str = ', '.join([f"C{c}" for c in communities])
            report += f"{i+1}. **{info['node']}** - Active in communities: {communities_str}\n"
        report += "\n"
    
    # Add structural roles
    if 'structural_roles' in community_results:
        roles = community_results['structural_roles']
        
        report += "## Structural Roles\n\n"
        
        for role, nodes in roles.items():
            report += f"### {role} Contributors\n\n"
            
            if len(nodes) > 10:
                report += f"*{len(nodes)} contributors with this role. Top 10 shown:*\n\n"
                nodes = nodes[:10]
            
            for node in nodes:
                report += f"- **{node}**\n"
            
            report += "\n"
    
    # Add recommendations
    report += "## Recommendations for Influencer Management\n\n"
    
    report += "1. **Recognize and Support Key Contributors**\n"
    report += "   - Acknowledge top contributors publicly\n"
    report += "   - Provide resources and support to prevent burnout\n\n"
    
    report += "2. **Bridge Knowledge Gaps**\n"
    report += "   - Encourage cross-community influencers to document interfaces\n"
    report += "   - Create formal knowledge transfer processes\n\n"
    
    report += "3. **Develop New Influencers**\n"
    report += "   - Identify potential new bridge and hub contributors\n"
    report += "   - Create mentorship relationships between established and emerging influencers\n\n"
    
    report += "4. **Reduce Bus Factor Risk**\n"
    report += "   - Distribute critical knowledge across multiple contributors\n"
    report += "   - Implement 'Shadow Maintainer' roles for bus factor reduction\n\n"
    
    report += "5. **Optimize Communication Channels**\n"
    report += "   - Ensure bridge contributors have effective ways to share information\n"
    report += "   - Create dedicated channels for cross-community coordination\n"
    
    # Save if requested
    if save_path:
        try:
            with open(f"{save_path}/influencer_report.md", 'w') as f:
                f.write(report)
        except Exception as e:
            print(f"Error saving influencer report: {e}")
    
    return report

def generate_community_health_report(community_results, temporal_results, save_path=None):
    """
    Generate a specialized report on community health and evolution.
    
    Parameters:
    - community_results: Results from community detection and influence analysis
    - temporal_results: Results from temporal evolution analysis
    - save_path: Path to save the report
    
    Returns:
    - Markdown report as a string
    """
    report = "# Community Health and Evolution Report\n\n"
    
    # Community structure overview
    if 'community_profiles' in community_results:
        profiles = community_results['community_profiles']
        
        report += "## Community Structure Overview\n\n"
        report += f"The project contains {len(profiles)} distinct contributor communities.\n\n"
        
        # Community statistics
        sizes = [p['size'] for p in profiles.values()]
        report += f"- **Largest community**: {max(sizes)} contributors\n"
        report += f"- **Smallest community**: {min(sizes)} contributors\n"
        report += f"- **Average community size**: {sum(sizes)/len(sizes):.1f} contributors\n\n"
        
        # Add table of communities
        report += "### Community Profiles\n\n"
        report += "| Community | Size | Density | Cohesion | Top Contributor |\n"
        report += "|-----------|------|---------|----------|----------------|\n"
        
        for comm_id, profile in sorted(profiles.items(), key=lambda x: x[1]['size'], reverse=True):
            top_contributor = profile.get('top_contributor', 'Unknown')
            report += f"| C{comm_id} | {profile['size']} | {profile.get('density', 'N/A'):.2f} | "
            report += f"{profile.get('cohesion', 'N/A'):.2f} | {top_contributor} |\n"
        
        report += "\n"
    
    # Temporal evolution
    if 'window_stats' in temporal_results:
        stats = temporal_results['window_stats']
        valid_stats = [s for s in stats if not s.get('too_small', False) and 'error' not in s]
        
        if len(valid_stats) >= 2:
            report += "## Community Evolution Over Time\n\n"
            
            # Extract time series data
            times = [s.get('label', f"Period {i+1}") for i, s in enumerate(valid_stats)]
            communities = [s.get('communities', 0) for s in valid_stats]
            nodes = [s.get('nodes', 0) for s in valid_stats]
            modularity = [s.get('modularity', 0) for s in valid_stats]
            
            # Overall trend
            first_period = valid_stats[0]
            last_period = valid_stats[-1]
            
            # Community count change
            comm_change = last_period.get('communities', 0) - first_period.get('communities', 0)
            comm_change_pct = (comm_change / first_period.get('communities', 1)) * 100
            
            report += f"Over the analyzed time periods, the number of communities has "
            if comm_change > 0:
                report += f"**increased by {comm_change_pct:.1f}%** ({comm_change} new communities).\n\n"
            elif comm_change < 0:
                report += f"**decreased by {abs(comm_change_pct):.1f}%** ({abs(comm_change)} fewer communities).\n\n"
            else:
                report += "**remained stable**.\n\n"
            
            # Contributor change
            node_change = last_period.get('nodes', 0) - first_period.get('nodes', 0)
            node_change_pct = (node_change / first_period.get('nodes', 1)) * 100
            
            report += f"The number of active contributors has "
            if node_change > 0:
                report += f"**increased by {node_change_pct:.1f}%** ({node_change} new contributors).\n\n"
            elif node_change < 0:
                report += f"**decreased by {abs(node_change_pct):.1f}%** ({abs(node_change)} fewer contributors).\n\n"
            else:
                report += "**remained stable**.\n\n"
            
            # Modularity change (community definition)
            mod_change = last_period.get('modularity', 0) - first_period.get('modularity', 0)
            
            report += f"Community definition has "
            if mod_change > 0.1:
                report += "**become much more distinct** over time.\n\n"
            elif mod_change > 0.05:
                report += "**become somewhat more distinct** over time.\n\n"
            elif mod_change < -0.1:
                report += "**become much less distinct** over time (communities are blending).\n\n"
            elif mod_change < -0.05:
                report += "**become somewhat less distinct** over time (communities are blending).\n\n"
            else:
                report += "**remained relatively stable** in distinctiveness.\n\n"
                
            # Key events and milestone analysis
            if 'key_events' in temporal_results:
                report += "### Key Events and Milestones\n\n"
                
                for event in temporal_results['key_events']:
                    report += f"- **{event['date']}**: {event['description']}\n"
                    if 'impact' in event:
                        report += f"  - *Impact*: {event['impact']}\n"
                
                report += "\n"
                
            # Community stability analysis
            if 'community_stability' in temporal_results:
                stability = temporal_results['community_stability']
                
                report += "### Community Stability Analysis\n\n"
                report += f"- **Most stable community**: Community {stability['most_stable']['id']} "
                report += f"(stability score: {stability['most_stable']['score']:.2f})\n"
                report += f"- **Most volatile community**: Community {stability['most_volatile']['id']} "
                report += f"(stability score: {stability['most_volatile']['score']:.2f})\n\n"
                
                # If we have community merge/split events
                if 'merge_split_events' in stability:
                    report += "#### Community Merges and Splits\n\n"
                    
                    for event in stability['merge_split_events']:
                        if event['type'] == 'merge':
                            report += f"- **{event['date']}**: Communities {', '.join([str(c) for c in event['source']])} "
                            report += f"merged into Community {event['destination']}\n"
                        elif event['type'] == 'split':
                            report += f"- **{event['date']}**: Community {event['source']} split into "
                            report += f"Communities {', '.join([str(c) for c in event['destination']])}\n"
                    
                    report += "\n"
    
    # Community health indicators
    report += "## Community Health Indicators\n\n"
    
    if 'health_metrics' in community_results:
        health = community_results['health_metrics']
        
        # Overall health score
        if 'overall_score' in health:
            overall = health['overall_score']
            report += f"### Overall Health Score: {overall:.1f}/10\n\n"
            
            if overall >= 8.5:
                report += "The project demonstrates **excellent community health** with balanced participation, "
                report += "sustainable growth, and effective knowledge sharing.\n\n"
            elif overall >= 7.0:
                report += "The project demonstrates **good community health** with generally positive trends in "
                report += "participation and knowledge sharing, though some areas could be improved.\n\n"
            elif overall >= 5.0:
                report += "The project demonstrates **moderate community health** with some positive aspects but "
                report += "significant areas needing attention.\n\n"
            else:
                report += "The project demonstrates **concerning community health** with multiple indicators "
                report += "suggesting potential sustainability risks.\n\n"
        
        # Specific health metrics
        report += "### Individual Health Metrics\n\n"
        report += "| Metric | Score | Interpretation |\n"
        report += "|--------|-------|----------------|\n"
        
        for metric, details in health.get('metrics', {}).items():
            if metric == 'overall_score':
                continue
                
            score = details.get('score', 'N/A')
            if isinstance(score, (int, float)):
                score_str = f"{score:.1f}/10"
            else:
                score_str = str(score)
                
            report += f"| {metric.replace('_', ' ').title()} | {score_str} | {details.get('interpretation', 'N/A')} |\n"
        
        report += "\n"
    
    # Recommendations for community health improvement
    report += "## Recommendations for Community Health Improvement\n\n"
    
    if 'recommendations' in community_results:
        for i, rec in enumerate(community_results['recommendations'], 1):
            report += f"### {i}. {rec['title']}\n\n"
            report += f"{rec['description']}\n\n"
            
            if 'actions' in rec:
                report += "**Suggested Actions:**\n\n"
                for action in rec['actions']:
                    report += f"- {action}\n"
                report += "\n"
                
            if 'expected_impact' in rec:
                report += f"**Expected Impact**: {rec['expected_impact']}\n\n"
    else:
        # Generic recommendations if specific ones aren't available
        report += "### 1. Improve Cross-Community Collaboration\n\n"
        report += "Enhance knowledge sharing between different community clusters by organizing cross-team events and projects.\n\n"
        report += "**Suggested Actions:**\n\n"
        report += "- Implement regular cross-community review sessions\n"
        report += "- Create shared documentation repositories\n"
        report += "- Assign 'diplomats' to bridge community gaps\n\n"
        
        report += "### 2. Strengthen Onboarding Process\n\n"
        report += "Develop a more structured onboarding process to help new contributors integrate into the community faster.\n\n"
        report += "**Suggested Actions:**\n\n"
        report += "- Create a mentorship program pairing new contributors with experienced ones\n"
        report += "- Identify and tag 'good first issues' for newcomers\n"
        report += "- Develop interactive onboarding documentation\n\n"
        
        report += "### 3. Reduce Dependency on Key Contributors\n\n"
        report += "Distribute knowledge and responsibilities to reduce the risk associated with high bus factor.\n\n"
        report += "**Suggested Actions:**\n\n"
        report += "- Implement pair programming or review practices\n"
        report += "- Document critical processes and knowledge areas\n"
        report += "- Rotate leadership roles for key community activities\n\n"
    
    # Save report if path is provided
    if save_path:
        try:
            with open(save_path+"community_health_report.txt", 'w') as f:
                f.write(report)
        except Exception as e:
            print(f"Error saving community health report: {e}")
    
    return report


def generate_governance_recommendations(analysis_results, save_path='out/'):
    """
    Generate specific governance recommendations based on network analysis.
    
    Parameters:
    - analysis_results: Combined results from various analyses
    - save_path: Path to save the recommendations
    
    Returns:
    - Markdown recommendations as a string
    """
    # Extract relevant metrics
    network_metrics = analysis_results.get('network_metrics', {})
    community_results = analysis_results.get('community_results', {})
    temporal_results = analysis_results.get('temporal_results', {})
    
    # Determine project size and maturity
    node_count = network_metrics.get('n_nodes', 0)
    if node_count < 20:
        size_category = "small"
    elif node_count < 100:
        size_category = "medium"
    else:
        size_category = "large"
    
    # Determine centralization level
    centralization = network_metrics.get('degree_centralization', 0.5)
    if centralization > 0.7:
        centralization_level = "highly centralized"
    elif centralization > 0.4:
        centralization_level = "moderately centralized"
    else:
        centralization_level = "decentralized"
    
    # Determine community structure
    modularity = network_metrics.get('modularity', 0.3)
    if modularity > 0.6:
        community_structure = "strong community structure"
    elif modularity > 0.3:
        community_structure = "moderate community structure"
    else:
        community_structure = "weak community structure"
    
    # Generate the report
    report = "# Project Governance Recommendations\n\n"
    report += f"Based on comprehensive network analysis of a **{size_category}** project with "
    report += f"**{centralization_level}** contributor structure and **{community_structure}**, "
    report += "the following governance recommendations are provided:\n\n"
    
    # Core governance model recommendation
    report += "## Recommended Governance Model\n\n"
    
    if size_category == "small" and centralization_level == "highly centralized":
        governance_model = "BDFL (Benevolent Dictator for Life)"
        description = "A single project leader makes final decisions while encouraging community input."
    elif size_category == "small" and centralization_level != "highly centralized":
        governance_model = "Consensus-Based"
        description = "Decisions are made through group consensus among all active contributors."
    elif size_category == "medium" and community_structure in ["weak community structure", "moderate community structure"]:
        governance_model = "Meritocratic"
        description = "Decision-making authority is earned through consistent quality contributions."
    elif size_category == "medium" and community_structure == "strong community structure":
        governance_model = "Federation with Central Coordination"
        description = "Semi-autonomous teams with a central coordination committee."
    elif size_category == "large" and centralization_level == "decentralized":
        governance_model = "Foundation Model"
        description = "A formal foundation provides governance structure with board members and clear processes."
    else:
        governance_model = "Council Governance"
        description = "Representative council from different stakeholder groups makes decisions."
    
    report += f"### {governance_model}\n\n"
    report += f"{description}\n\n"
    
    # Specific governance structures
    report += "## Governance Structure Implementation\n\n"
    
    # Decision-making process
    report += "### Decision-Making Process\n\n"
    
    if governance_model == "BDFL":
        report += "- Project lead has final decision authority\n"
        report += "- Regular community feedback sessions\n"
        report += "- Transparent decision logs with reasoning\n"
        report += "- Clear escalation path for significant disagreements\n\n"
    elif governance_model == "Consensus-Based":
        report += "- Asynchronous voting on all significant changes\n"
        report += "- Required approval from >70% of active contributors\n"
        report += "- Time-boxed discussion periods (1-2 weeks)\n"
        report += "- Conflict resolution process for deadlocks\n\n"
    elif governance_model == "Meritocratic":
        report += "- Tiered voting rights based on contribution history\n"
        report += "- Core team approval required for architectural changes\n"
        report += "- Open discussion period followed by maintainer decision\n"
        report += "- Path for non-maintainers to propose significant changes\n\n"
    elif governance_model == "Federation":
        report += "- Team autonomy for team-specific decisions\n"
        report += "- Central committee for cross-cutting concerns\n"
        report += "- Regular inter-team coordination meetings\n"
        report += "- Written team charters defining boundaries\n\n"
    elif governance_model == "Foundation Model":
        report += "- Formal board elections (annual/biannual)\n"
        report += "- Published governance documents and bylaws\n"
        report += "- Working groups for specific areas\n"
        report += "- Transparent meeting minutes and decision records\n\n"
    else:  # Council
        report += "- Representative council from each community cluster\n"
        report += "- Rotating council chair\n"
        report += "- Published decision-making framework\n"
        report += "- Regular open meetings with community\n\n"
    
    # Roles and responsibilities
    report += "### Roles and Responsibilities\n\n"
    
    # Core roles that most projects need
    report += "#### Core Project Roles\n\n"
    report += "| Role | Responsibilities | Selection Process |\n"
    report += "|------|------------------|-------------------|\n"
    
    if size_category == "small":
        report += "| Project Lead | Final decision authority, roadmap planning | Project founder or elected |\n"
        report += "| Maintainer | Code review, merge approval | Appointed by Project Lead |\n"
        report += "| Contributor | Submit patches, report issues | Self-selected |\n"
    elif size_category == "medium":
        report += "| Core Team | Architecture decisions, roadmap planning | Meritocracy-based promotion |\n"
        report += "| Maintainer | Subsystem ownership, code review | Appointed by Core Team |\n"
        report += "| Regular Contributor | Feature development, bug fixes | Self-selected |\n"
        report += "| Community Manager | Outreach, community health | Appointed or elected |\n"
    else:  # large
        report += "| Board/Council | Strategic direction, conflict resolution | Elected by community |\n"
        report += "| Working Group Lead | Feature area leadership | Appointed by Board |\n"
        report += "| Maintainer | Component ownership, review | Meritocracy-based promotion |\n"
        report += "| Community Manager | Onboarding, events, metrics | Hired or elected |\n"
        report += "| Documentation Lead | Documentation strategy | Appointed or elected |\n"
    
    report += "\n"
    
    # Project-specific role recommendations based on network structure
    if 'structural_roles' in community_results:
        report += "#### Project-Specific Role Recommendations\n\n"
        
        for role in community_results.get('structural_roles', []):
            report += f"- **{role['name']}**: {role['description']}\n"
            if 'recommended_contributors' in role and role['recommended_contributors']:
                report += f"  - *Suggested contributors*: {', '.join(role['recommended_contributors'][:3])}\n"
        report += "\n"
    
    # Communication channels
    report += "### Communication Structure\n\n"
    
    report += "#### Recommended Communication Channels\n\n"
    report += "| Channel | Purpose | Frequency |\n"
    report += "|---------|---------|-----------|"
    
    if size_category == "small":
        report += "\n| Team chat | Daily coordination | Ongoing |\n"
        report += "| Video standup | Progress updates | Weekly |\n"
        report += "| Issue discussions | Feature planning | As needed |\n"
    elif size_category == "medium":
        report += "\n| Team chat | Daily coordination | Ongoing |\n"
        report += "| Maintainer meeting | Strategic planning | Biweekly |\n" 
        report += "| Community call | Updates, demos | Monthly |\n"
        report += "| Mailing list | Major announcements | As needed |\n"
    else:  # large
        report += "\n| Team chat (per working group) | Daily coordination | Ongoing |\n"
        report += "| Council meeting | Governance, cross-cutting issues | Biweekly |\n"
        report += "| All-hands call | Project-wide updates | Monthly |\n"
        report += "| Working group meetings | Feature planning | Weekly |\n"
        report += "| Blog/newsletter | Community updates | Monthly |\n"
    
    report += "\n"
    
    # Cross-community collaboration
    if community_structure == "strong community structure":
        report += "#### Cross-Community Collaboration Mechanisms\n\n"
        report += "- Designated 'bridge' contributors between communities\n"
        report += "- Regular joint planning sessions across communities\n"
        report += "- Rotation program for contributors to work in different areas\n"
        report += "- Cross-community review requirements for architectural changes\n\n"
    
    # Conflict resolution
    report += "### Conflict Resolution Process\n\n"
    report += "1. **Direct discussion**: Involved parties attempt to resolve directly\n"
    report += "2. **Maintainer mediation**: Area maintainer facilitates resolution\n"
    
    if size_category == "small":
        report += "3. **Project lead decision**: Final resolution if needed\n\n"
    elif governance_model in ["Council Governance", "Foundation Model"]:
        report += "3. **Working group escalation**: Brought to relevant working group\n"
        report += "4. **Council/board decision**: Final resolution if needed\n\n"
    else:
        report += "3. **Community discussion**: Time-boxed public discussion\n"
        report += "4. **Core team vote**: Final resolution if needed\n\n"
    
    # Contribution workflow
    report += "### Contribution Workflow\n\n"
    
    report += "#### Recommended Process\n\n"
    report += "1. **Issue creation** - Document the problem/feature\n"
    report += "2. **Discussion** - Gather feedback on approach\n"
    report += "3. **Implementation** - Create the change with tests\n"
    report += "4. **Review** - Peer review and approval\n"
    report += "5. **Integration** - Merge and release planning\n\n"
    
    report += "#### Review Requirements\n\n"
    
    if centralization_level == "highly centralized":
        report += "- At least one approval from a core maintainer\n"
        report += "- Project lead review for architectural changes\n"
    elif size_category == "small":
        report += "- At least one approval from any team member\n"
        report += "- All team review for major changes\n"
    else:
        report += "- At least two approvals including one from component maintainer\n"
        report += "- Architectural council review for cross-cutting concerns\n"
    
    report += "\n"
    
    # Implementation plan
    if 'current_governance' in analysis_results:
        current = analysis_results['current_governance']
        report += "## Transition Plan\n\n"
        report += f"To transition from the current {current['model']} model to the recommended "
        report += f"{governance_model} approach:\n\n"
        
        report += "1. **Documentation Phase** (1 month)\n"
        report += "   - Draft governance documents\n"
        report += "   - Gather community feedback\n"
        report += "   - Finalize initial structure\n\n"
        
        report += "2. **Trial Period** (3 months)\n"
        report += "   - Implement new processes on trial basis\n"
        report += "   - Regular retrospectives and adjustments\n"
        report += "   - Training for new roles\n\n"
        
        report += "3. **Full Implementation** (ongoing)\n"
        report += "   - Official governance transition\n"
        report += "   - Regular governance retrospectives\n"
        report += "   - Annual governance review\n\n"
    
    # Save report if path is provided
    if save_path:
        try:
            with open(save_path+"governance_recommendations", 'w') as f:
                f.write(report)
        except Exception as e:
            print(f"Error saving governance recommendations: {e}")
    
    return report


def visualize_influencer_network(G, metrics, communities=None, top_n=20):
    """
    Create an advanced visualization of top influencers and their relationships.
    
    Parameters:
    - G: NetworkX graph object
    - metrics: Dictionary of node metrics
    - communities: Community assignment dictionary
    - top_n: Number of top influencers to highlight
    
    Returns:
    - Figure object
    """
    # Create a copy of the graph to visualize
    G_viz = G.copy()
    
    # Identify top influencers by composite score
    if metrics and 'influencer_score' in next(iter(metrics.values()), {}):
        top_influencers = sorted(
            G.nodes(), 
            key=lambda n: metrics[n].get('influencer_score', 0), 
            reverse=True
        )[:top_n]
    else:
        # Fall back to betweenness if no influencer score
        top_influencers = sorted(
            G.nodes(), 
            key=lambda n: metrics[n].get('betweenness', 0), 
            reverse=True
        )[:top_n]
    
    # Set up node attributes for visualization
    node_size = []
    node_color = []
    alpha = []
    labels = {}
    
    for node in G_viz.nodes():
        # Size based on influencer score or betweenness
        if 'influencer_score' in metrics[node]:
            size = 50 + 500 * metrics[node]['influencer_score']
        else:
            size = 50 + 500 * metrics[node].get('betweenness', 0)
        
        # Set size
        node_size.append(size)
        
        # Set color based on community
        if communities and node in communities:
            color_idx = communities[node] % len(PALETTE)
            node_color.append(PALETTE[color_idx])
        else:
            node_color.append('#CCCCCC')  # Default gray
            
        # Set visibility
        if node in top_influencers:
            alpha.append(0.9)  # High visibility for top influencers
            labels[node] = node  # Only label top influencers
        else:
            alpha.append(0.2)  # Reduced visibility for other nodes
    
    # Create figure
    plt.figure(figsize=(16, 12))
    
    # Use a force-directed layout focused on top influencers
    pos = nx.spring_layout(G_viz, k=0.3, iterations=50, seed=42)
    
    # Draw edges with transparency
    nx.draw_networkx_edges(G_viz, pos, alpha=0.1, width=0.5)
    
    # Draw nodes
    nx.draw_networkx_nodes(
        G_viz, pos,
        node_size=node_size,
        node_color=node_color,
        alpha=alpha
    )
    
    # Draw labels for top influencers only
    nx.draw_networkx_labels(
        G_viz, pos,
        labels=labels,
        font_size=10,
        font_weight='bold',
        font_color='black',
        bbox=dict(facecolor='white', edgecolor='none', alpha=0.7, pad=0.5)
    )
    
    # Add legend for top 5 influencers
    top_5 = top_influencers[:5]
    legend_elements = []
    
    for i, node in enumerate(top_5):
        if communities and node in communities:
            color_idx = communities[node] % len(PALETTE)
            color = PALETTE[color_idx]
        else:
            color = '#CCCCCC'
        
        score = metrics[node].get('influencer_score', metrics[node].get('betweenness', 0))
        legend_elements.append(
            mpatches.Patch(
                color=color,
                label=f"{i+1}. {node} (score: {score:.2f})"
            )
        )
    
    plt.legend(
        handles=legend_elements,
        title="Top 5 Influencers",
        loc='upper right',
        fontsize=10
    )
    
    # Set title and layout
    plt.title("Key Influencers in Collaboration Network", fontsize=16)
    plt.tight_layout()
    
    return plt.gcf()

```
